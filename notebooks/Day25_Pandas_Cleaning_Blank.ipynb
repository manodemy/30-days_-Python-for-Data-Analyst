<h2 style="color: #000000ff; background-color: #02fff2ff; border-left: 6px solid #0c0c0cff; padding: 12px 15px; margin-top: 40px; border-radius: 4px;">
    <b>1. Handling Missing Data : <span style="color: #070707ff; font-size: 0.85em; font-weight: normal;">The Reality of Data</span></b>
</h2>

<div style="border-left: 4px solid #4DB6AC; padding-left: 15px; margin: 15px 0; color: #CCCCCC;">
    <b>🔍 What is it?</b><br>
    Real-world data is messy. <b>Missing values</b> are represented as <code>NaN</code> (Not a Number) or <code>None</code>. Pandas provides methods to detect (<code>isna()</code>), remove (<code>dropna()</code>), or replace (<code>fillna()</code>) these missing values.
</div>

| Method | Action | Common Use |
| :--- | :--- | :--- |
| `df.isna()` | Returns Boolean DF | Identifying where data is missing |
| `df.dropna()` | Drops rows/cols | Removing rows where critical data is null |
| `df.fillna(val)` | Replaces NaNs | Replacing missing ages with the median age |

<h3 style="color: #64B5F6; background-color: #0F1A2A; padding: 8px 12px; border-radius: 4px; margin-top: 25px;"><b>💼 Why Data Analysts Care</b></h3>

* **Data Quality**: Machine Learning models will crash if you feed them `NaN` values. You must handle them first!
* **Imputation**: Filling missing salaries with the average salary of the employee's department

<h3 style="color: #FFB74D; background-color: #2B1A0D; padding: 8px 12px; border-radius: 4px; margin-top: 25px;"><b>⚠️ Blindly Dropping</b></h3>

<div style="border-left: 4px solid #FFB74D; padding-left: 15px; margin: 15px 0; color: #CCCCCC;">
    Never run <code>df.dropna()</code> without thinking. If 30% of your data has a missing 'Phone Number', dropping those rows destroys 30% of your valid 'Email' and 'Name' data. Drop the column, or fill it with 'Unknown'.
</div>

````python
import pandas as pd
import numpy as np

df = pd.DataFrame({
    'Name': ['Alice', 'Bob', 'Charlie', 'David'],
    'Age': [25, np.nan, 35, 40],
    'Salary': [50000, 60000, np.nan, 80000]
})

print("Original:\n", df)

# 1. Count missing values per column
print("\nMissing Counts:\n", df.isna().sum())

# 2. Drop rows with ANY missing data
print("\nDropped Rows:\n", df.dropna())

# 3. Fill missing Age with Mean, Salary with 0
clean_df = df.fillna({'Age': df['Age'].mean(), 'Salary': 0})
print("\nFilled NaNs:\n", clean_df)
````

<h2 style="color: #000000ff; background-color: #02fff2ff; border-left: 6px solid #0c0c0cff; padding: 12px 15px; margin-top: 40px; border-radius: 4px;">
    <b>2. Handling Duplicates : <span style="color: #070707ff; font-size: 0.85em; font-weight: normal;">Deduplication</span></b>
</h2>

<div style="border-left: 4px solid #4DB6AC; padding-left: 15px; margin: 15px 0; color: #CCCCCC;">
    <b>🔍 What is it?</b><br>
    Duplicate rows can skew aggregations and ruin analysis. <code>df.duplicated()</code> returns a boolean mask of duplicate rows, and <code>df.drop_duplicates()</code> removes them. You can specify whether to keep the 'first' occurrence, the 'last', or False (drop all duplicates completely).
</div>

```python
# Keep only the first occurrence of a User_ID
df.drop_duplicates(subset=['User_ID'], keep='first')
```

<h3 style="color: #64B5F6; background-color: #0F1A2A; padding: 8px 12px; border-radius: 4px; margin-top: 25px;"><b>💼 Why Data Analysts Care</b></h3>

* **ETL Pipelines**: When pulling data daily, you might accidentally pull the same transactions twice. Deduplication fixes this
* **Latest Records**: Sorting by 'Update_Date' and keeping the 'last' duplicate ensures you only have the most recent data for each user

<h3 style="color: #81C784; background-color: #102414; padding: 8px 12px; border-radius: 4px; margin-top: 25px;"><b>🧠 Pro Tip</b></h3>

<div style="border-left: 4px solid #81C784; padding-left: 15px; margin: 15px 0; color: #CCCCCC;">
    Always sort your DataFrame BEFORE dropping duplicates if you want to keep the highest/lowest/newest value. E.g., Sort by `Date` ascending, then `drop_duplicates(keep='last')`.
</div>

````python
import pandas as pd

df = pd.DataFrame({
    'ID': [1, 2, 2, 3, 1],
    'Name': ['Alice', 'Bob', 'Bob', 'Charlie', 'Alice'],
    'Update_Day': [1, 1, 2, 1, 3]
})

print("Original with duplicates:\n", df)

# Drop exact identical rows (None match exactly here because Update_Day differs)
print("\nExact Duplicates:\n", df.drop_duplicates())

# Drop based on ID, keeping the latest Update_Day
# 1. Sort by Day
df_sorted = df.sort_values(by='Update_Day')
# 2. Drop duplicates based on ID, keeping the LAST one
latest_df = df_sorted.drop_duplicates(subset=['ID'], keep='last')
print("\nLatest Unique Users:\n", latest_df.sort_index())
````

<h2 style="color: #000000ff; background-color: #02fff2ff; border-left: 6px solid #0c0c0cff; padding: 12px 15px; margin-top: 40px; border-radius: 4px;">
    <b>3. Data Type Conversion : <span style="color: #070707ff; font-size: 0.85em; font-weight: normal;">Fixing Bad Formatting</span></b>
</h2>

<div style="border-left: 4px solid #4DB6AC; padding-left: 15px; margin: 15px 0; color: #CCCCCC;">
    <b>🔍 What is it?</b><br>
    When reading from CSVs, numbers might be loaded as strings (e.g., `'1,000'`). You cannot perform math on strings! The <code>.astype()</code> method converts columns to proper types. For strings requiring cleaning, the <code>.str</code> accessor is your best friend.
</div>

| Method | Action | Example |
| :--- | :--- | :--- |
| `astype(type)` | Cast to type | `df['Age'].astype(int)` |
| `pd.to_numeric()` | Safe casting | `pd.to_numeric(df['Price'], errors='coerce')` |
| `.str.replace()`| String cleanup| `df['Price'].str.replace('$', '')` |

<h3 style="color: #64B5F6; background-color: #0F1A2A; padding: 8px 12px; border-radius: 4px; margin-top: 25px;"><b>💼 Why Data Analysts Care</b></h3>

* **Currency Parsing**: Converting `'$1,200.50'` into the float `1200.50` so you can sum it
* **Memory Optimization**: Downcasting an `int64` column to `int8` to save 8x the RAM on a billion rows

<h3 style="color: #81C784; background-color: #102414; padding: 8px 12px; border-radius: 4px; margin-top: 25px;"><b>🧠 Pro Tip</b></h3>

<div style="border-left: 4px solid #81C784; padding-left: 15px; margin: 15px 0; color: #CCCCCC;">
    If <code>.astype(float)</code> fails because of a weird string like 'N/A', use <code>pd.to_numeric(col, errors='coerce')</code>. It will force the bad strings into <code>NaN</code> so you can proceed!
</div>

````python
import pandas as pd

df = pd.DataFrame({
    'Item': ['A', 'B', 'C'],
    'Price': ['$10', '$20', 'Error']
})
print("Original types:\n", df.dtypes)

# 1. Clean the string (Remove $)
df['Clean_Price'] = df['Price'].str.replace('$', '', regex=False)

# 2. Convert to numeric safely (Forces 'Error' to NaN)
df['Float_Price'] = pd.to_numeric(df['Clean_Price'], errors='coerce')

print("\nCleaned DataFrame:\n", df)
print("\nNew types:\n", df.dtypes)
````
