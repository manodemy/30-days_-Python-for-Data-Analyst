<h3 style="color: #FFB74D; background-color: #2B1A0D; padding: 8px 12px; border-radius: 4px; margin-top: 25px;"><b>⚠️ Over-engineering</b></h3>

<div style="border-left: 4px solid #FFB74D; padding-left: 15px; margin: 15px 0; color: #CCCCCC;">
    Don't use OOP if a simple dictionary and a function will do the job. Python supports both functional and OOP paradigms. Use OOP when you need to manage complex <b>state</b>.
</div>

````python
class Customer:
    # A basic class
    pass

# Create two distinct objects
c1 = Customer()
c2 = Customer()

# Add attributes dynamically (not recommended, but possible)
c1.name = "Alice"
c2.name = "Bob"

print(f"C1: {c1.name}")
print(f"C2: {c2.name}")
print(c1 == c2) # False, they are distinct objects in memory
````

<h2 style="color: #000000ff; background-color: #02fff2ff; border-left: 6px solid #0c0c0cff; padding: 12px 15px; margin-top: 40px; border-radius: 4px;">
    <b>2. __init__ and self : <span style="color: #070707ff; font-size: 0.85em; font-weight: normal;">The Constructor</span></b>
</h2>

<div style="border-left: 4px solid #4DB6AC; padding-left: 15px; margin: 15px 0; color: #CCCCCC;">
    <b>🔍 What is it?</b><br>
    To set up an object properly when it is created, we use the <b>constructor method</b>: <code>__init__</code>. The first parameter of EVERY method must be <code>self</code>, which refers to the specific object being manipulated.
</div>

```python
class User:
    def __init__(self, name, age):
        self.name = name  # Attribute assignment
        self.age = age

alice = User('Alice', 25) # __init__ is called automatically
```

<h3 style="color: #64B5F6; background-color: #0F1A2A; padding: 8px 12px; border-radius: 4px; margin-top: 25px;"><b>💼 Why Data Analysts Care</b></h3>

* **Initialization**: Ensuring a database connection object immediately connects when instantiated
* **State Management**: Tracking variables like `self.is_logged_in` across different method calls

<h3 style="color: #81C784; background-color: #102414; padding: 8px 12px; border-radius: 4px; margin-top: 25px;"><b>🧠 Pro Tip</b></h3>

<div style="border-left: 4px solid #81C784; padding-left: 15px; margin: 15px 0; color: #CCCCCC;">
    Dunder methods (like <code>__init__</code>, with double underscores) are 'magic methods' that Python calls automatically behind the scenes. You rarely call them directly.
</div>

````python
class Product:
    # The constructor
    def __init__(self, name, price):
        self.name = name
        self.price = price
        self.in_stock = True  # Default state
        
    # An instance method
    def apply_discount(self, percent):
        self.price = self.price * (1 - percent)
        print(f"{self.name} discounted to {self.price}")

# Instantiate and use
laptop = Product("MacBook", 1000)
laptop.apply_discount(0.20)
print(f"In stock? {laptop.in_stock}")
````

<h2 style="color: #000000ff; background-color: #02fff2ff; border-left: 6px solid #0c0c0cff; padding: 12px 15px; margin-top: 40px; border-radius: 4px;">
    <b>3. Inheritance : <span style="color: #070707ff; font-size: 0.85em; font-weight: normal;">Reusing Class Logic</span></b>
</h2>

<div style="border-left: 4px solid #4DB6AC; padding-left: 15px; margin: 15px 0; color: #CCCCCC;">
    <b>🔍 What is it?</b><br>
    <b>Inheritance</b> allows a new Class (Child) to adopt the attributes and methods of an existing Class (Parent). This promotes code reuse. You use <code>super()</code> to call methods from the Parent class.
</div>

```python
class Employee:
    def work(self): print('Working')

class Manager(Employee): # Inherits from Employee
    def meeting(self): print('In a meeting')
```

<h3 style="color: #64B5F6; background-color: #0F1A2A; padding: 8px 12px; border-radius: 4px; margin-top: 25px;"><b>💼 Why Data Analysts Care</b></h3>

* **Code Reuse**: Creating `CSVReader` and `JSONReader` that both inherit from a base `DataReader` class
* **Specialization**: Creating specific exception types that inherit from `Exception`

<h3 style="color: #FFB74D; background-color: #2B1A0D; padding: 8px 12px; border-radius: 4px; margin-top: 25px;"><b>⚠️ Deep Hierarchies</b></h3>

<div style="border-left: 4px solid #FFB74D; padding-left: 15px; margin: 15px 0; color: #CCCCCC;">
    Avoid creating deeply nested inheritance chains (A inherits B inherits C inherits D). It makes code impossible to debug. Prefer 'Composition' over deep Inheritance.
</div>

````python
class Animal:
    def __init__(self, name):
        self.name = name
    def speak(self):
        return "..."

# Dog inherits from Animal
class Dog(Animal):
    def speak(self):
        return "Woof!"

# Cat inherits from Animal, and adds a new attribute
class Cat(Animal):
    def __init__(self, name, color):
        super().__init__(name)  # Call parent constructor
        self.color = color
        
    def speak(self):
        return "Meow!"

d = Dog("Rex")
c = Cat("Luna", "Black")
print(f"{d.name}: {d.speak()}")
print(f"{c.color} cat {c.name}: {c.speak()}")
````
