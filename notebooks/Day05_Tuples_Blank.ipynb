<h2 style="color: #000000ff; background-color: #02fff2ff; border-left: 6px solid #0c0c0cff; padding: 12px 15px; margin-top: 40px; border-radius: 4px;">
    <b>1. Tuple Creation & Basics : <span style="color: #070707ff; font-size: 0.85em; font-weight: normal;">Immutable Ordered Sequences</span></b>
</h2>

<div style="border-left: 4px solid #4DB6AC; padding-left: 15px; margin: 15px 0; color: #CCCCCC;">
    <b>🔍 What is it?</b><br>
    A tuple is an <b>ordered, immutable collection</b> created with parentheses <code>()</code> or just commas. Once created, elements <b>cannot be added, removed, or changed</b>. This immutability makes tuples faster, hashable, and safer than lists.
</div>

| Creation Method | Example | Note |
| :--- | :--- | :--- |
| Parentheses | `(1, 2, 3)` | Standard |
| Without parens | `1, 2, 3` | Comma creates tuple |
| Single element | `(42,)` | Trailing comma required! |
| `tuple()` | `tuple([1,2,3])` | Convert from iterable |
| Empty | `()` or `tuple()` | Empty tuple |

<h3 style="color: #64B5F6; background-color: #0F1A2A; padding: 8px 12px; border-radius: 4px; margin-top: 25px;"><b>💼 Why Data Analysts Care</b></h3>

* **Database records**: Query results are returned as tuples — immutable rows
* **Function returns**: Functions return multiple values as tuples: `return x, y`
* **Configuration**: Immutable settings that should not be accidentally modified

<h3 style="color: #FFB74D; background-color: #2B1A0D; padding: 8px 12px; border-radius: 4px; margin-top: 25px;"><b>⚠️ Single Element Tuple</b></h3>

<div style="border-left: 4px solid #FFB74D; padding-left: 15px; margin: 15px 0; color: #CCCCCC;">
    <code>(42)</code> is just an integer in parentheses. You need a trailing comma: <code>(42,)</code> to create a single-element tuple.
</div>

<h3 style="color: #81C784; background-color: #102414; padding: 8px 12px; border-radius: 4px; margin-top: 25px;"><b>🧠 Pro Tip</b></h3>

<div style="border-left: 4px solid #81C784; padding-left: 15px; margin: 15px 0; color: #CCCCCC;">
    Tuples use <b>less memory</b> than lists and are <b>faster</b> to create and access. Use them for fixed collections.
</div>

````python
# Creation methods
coords = (10, 20)
rgb = 255, 128, 0       # Parens optional
single = (42,)           # Trailing comma!
empty = ()
from_list = tuple([1, 2, 3])

print(f'coords: {coords}, type: {type(coords)}')
print(f'rgb: {rgb}')
print(f'single: {single}, type: {type(single)}')
print(f'(42) type: {type((42))}')  # int, not tuple!

# Immutability
import sys
lst = [1, 2, 3]
tpl = (1, 2, 3)
print(f'List size: {sys.getsizeof(lst)} bytes')
print(f'Tuple size: {sys.getsizeof(tpl)} bytes')
````

<h2 style="color: #000000ff; background-color: #02fff2ff; border-left: 6px solid #0c0c0cff; padding: 12px 15px; margin-top: 40px; border-radius: 4px;">
    <b>2. Indexing, Slicing & Methods : <span style="color: #070707ff; font-size: 0.85em; font-weight: normal;">Accessing Tuple Data</span></b>
</h2>

<div style="border-left: 4px solid #4DB6AC; padding-left: 15px; margin: 15px 0; color: #CCCCCC;">
    <b>🔍 What is it?</b><br>
    Tuples support the same indexing and slicing as lists and strings. Since tuples are immutable, they have only <b>two methods</b>: <code>.count()</code> and <code>.index()</code>. No append, remove, or sort.
</div>

| Operation | Syntax | Description |
| :--- | :--- | :--- |
| Access | `t[0]`, `t[-1]` | By index |
| Slice | `t[1:3]` | Returns new tuple |
| Count | `t.count(x)` | Occurrences of x |
| Index | `t.index(x)` | First position of x |
| Length | `len(t)` | Number of elements |
| Membership | `x in t` | Check presence |

<h3 style="color: #64B5F6; background-color: #0F1A2A; padding: 8px 12px; border-radius: 4px; margin-top: 25px;"><b>💼 Why Data Analysts Care</b></h3>

* **Record access**: Access database row fields by position: `row[0]` for ID, `row[1]` for name
* **Data validation**: `.count()` to verify expected occurrences in fixed datasets
* **Slicing subsets**: Extract specific fields from structured records

<h3 style="color: #81C784; background-color: #102414; padding: 8px 12px; border-radius: 4px; margin-top: 25px;"><b>🧠 Pro Tip</b></h3>

<div style="border-left: 4px solid #81C784; padding-left: 15px; margin: 15px 0; color: #CCCCCC;">
    Tuples support all the same built-in functions as lists: <code>len()</code>, <code>min()</code>, <code>max()</code>, <code>sum()</code>, <code>sorted()</code> (returns a list).
</div>

````python
record = ('Alice', 28, 'Engineer', 'London', 85000)

# Indexing
print(f'Name: {record[0]}')
print(f'City: {record[-2]}')

# Slicing
print(f'Name+Age: {record[:2]}')

# Methods
scores = (85, 92, 78, 85, 95, 85)
print(f'Count of 85: {scores.count(85)}')  # 3
print(f'Index of 92: {scores.index(92)}')  # 1

# Built-in functions work
print(f'Min: {min(scores)}, Max: {max(scores)}')
print(f'Sorted (returns list): {sorted(scores)}')
````

<h2 style="color: #000000ff; background-color: #02fff2ff; border-left: 6px solid #0c0c0cff; padding: 12px 15px; margin-top: 40px; border-radius: 4px;">
    <b>3. Packing & Unpacking : <span style="color: #070707ff; font-size: 0.85em; font-weight: normal;">Elegant Data Extraction</span></b>
</h2>

<div style="border-left: 4px solid #4DB6AC; padding-left: 15px; margin: 15px 0; color: #CCCCCC;">
    <b>🔍 What is it?</b><br>
    <b>Packing</b> combines values into a tuple. <b>Unpacking</b> extracts tuple elements into individual variables. Python also supports <b>extended unpacking</b> with <code>*</code> to capture remaining elements.
</div>

```python
# Packing
point = 10, 20, 30

# Unpacking
x, y, z = point

# Extended unpacking
first, *rest = [1, 2, 3, 4, 5]  # first=1, rest=[2,3,4,5]
first, *mid, last = [1, 2, 3, 4, 5]  # first=1, mid=[2,3,4], last=5
```

<h3 style="color: #64B5F6; background-color: #0F1A2A; padding: 8px 12px; border-radius: 4px; margin-top: 25px;"><b>💼 Why Data Analysts Care</b></h3>

* **Multiple returns**: `mean, std = compute_stats(data)` — unpack function results
* **Swap variables**: `a, b = b, a` — Pythonic swap without temp variable
* **Iteration**: `for name, age in pairs:` — unpack in loops
* **CSV parsing**: `date, *values = row.split(',')` — separate header from data

<h3 style="color: #FFB74D; background-color: #2B1A0D; padding: 8px 12px; border-radius: 4px; margin-top: 25px;"><b>⚠️ Mismatched Unpacking</b></h3>

<div style="border-left: 4px solid #FFB74D; padding-left: 15px; margin: 15px 0; color: #CCCCCC;">
    Unpacking requires exact count match: <code>a, b = (1, 2, 3)</code> raises <code>ValueError</code>. Use <code>*</code> to absorb extras: <code>a, b, *rest = (1, 2, 3)</code>.
</div>

````python
# Packing
coord = 3.5, 7.2, 1.0
print(f'Packed: {coord}')

# Unpacking
x, y, z = coord
print(f'x={x}, y={y}, z={z}')

# Swap — most Pythonic pattern
a, b = 10, 20
a, b = b, a
print(f'Swapped: a={a}, b={b}')

# Extended unpacking
first, *rest = [1, 2, 3, 4, 5]
print(f'First: {first}, Rest: {rest}')

head, *body, tail = 'ABCDE'
print(f'Head: {head}, Body: {body}, Tail: {tail}')

# In loops
students = [('Alice', 95), ('Bob', 82), ('Charlie', 91)]
for name, score in students:
    print(f'{name}: {score}')
````

<h2 style="color: #000000ff; background-color: #02fff2ff; border-left: 6px solid #0c0c0cff; padding: 12px 15px; margin-top: 40px; border-radius: 4px;">
    <b>4. Named Tuples : <span style="color: #070707ff; font-size: 0.85em; font-weight: normal;">Self-Documenting Records</span></b>
</h2>

<div style="border-left: 4px solid #4DB6AC; padding-left: 15px; margin: 15px 0; color: #CCCCCC;">
    <b>🔍 What is it?</b><br>
    <code>collections.namedtuple</code> creates tuple subclasses with <b>named fields</b>. You get the immutability and performance of tuples, plus the readability of accessing fields by name instead of index.
</div>

```python
from collections import namedtuple
Employee = namedtuple('Employee', ['name', 'age', 'dept'])
emp = Employee('Alice', 30, 'Engineering')
print(emp.name)  # Alice (instead of emp[0])
```

<h3 style="color: #64B5F6; background-color: #0F1A2A; padding: 8px 12px; border-radius: 4px; margin-top: 25px;"><b>💼 Why Data Analysts Care</b></h3>

* **Database rows**: Map query results to named fields for readable access
* **Config objects**: Immutable configuration with meaningful attribute names
* **API responses**: Structure API data with clear field names
* **CSV records**: Parse CSV rows into typed, named records

<h3 style="color: #81C784; background-color: #102414; padding: 8px 12px; border-radius: 4px; margin-top: 25px;"><b>🧠 Pro Tip</b></h3>

<div style="border-left: 4px solid #81C784; padding-left: 15px; margin: 15px 0; color: #CCCCCC;">
    Named tuples support <code>._asdict()</code> to convert to dict, <code>._replace()</code> to create modified copies, and <code>._fields</code> to list field names.
</div>

````python
from collections import namedtuple

# Define a record type
Employee = namedtuple('Employee', ['name', 'age', 'dept', 'salary'])

# Create instances
alice = Employee('Alice', 30, 'Engineering', 95000)
bob = Employee(name='Bob', age=25, dept='Marketing', salary=72000)

# Access by name (readable!)
print(f'{alice.name} in {alice.dept}: ${alice.salary:,}')

# Still works with indexing
print(f'Name: {alice[0]}, Age: {alice[1]}')

# Useful methods
print(f'Fields: {Employee._fields}')
print(f'As dict: {alice._asdict()}')

# Create modified copy
senior_alice = alice._replace(salary=110000)
print(f'Promoted: {senior_alice}')
````

<h2 style="color: #000000ff; background-color: #02fff2ff; border-left: 6px solid #0c0c0cff; padding: 12px 15px; margin-top: 40px; border-radius: 4px;">
    <b>5. Tuples vs Lists : <span style="color: #070707ff; font-size: 0.85em; font-weight: normal;">When to Use Which</span></b>
</h2>

<div style="border-left: 4px solid #4DB6AC; padding-left: 15px; margin: 15px 0; color: #CCCCCC;">
    <b>🔍 What is it?</b><br>
    The choice between tuple and list depends on your intent: use <b>tuples for fixed, heterogeneous records</b> (like database rows) and <b>lists for variable-length, homogeneous collections</b> (like a series of measurements). Tuples signal "this data should not change."
</div>

| Feature | Tuple | List |
| :--- | :--- | :--- |
| Mutable? | No | Yes |
| Speed | Faster | Slower |
| Memory | Less | More |
| Hashable? | Yes (if contents are) | No |
| Dict key? | Yes | No |
| Methods | 2 (count, index) | 11+ |
| Use for | Fixed records | Dynamic collections |

<h3 style="color: #64B5F6; background-color: #0F1A2A; padding: 8px 12px; border-radius: 4px; margin-top: 25px;"><b>💼 Why Data Analysts Care</b></h3>

* **Dict keys**: Tuples can be dictionary keys (hashable): `{(lat, lon): 'city'}`
* **Set elements**: Tuples can be added to sets; lists cannot
* **Thread safety**: Immutable tuples are inherently thread-safe
* **Performance**: Tuple creation is ~5-10x faster than list creation

<h3 style="color: #FFB74D; background-color: #2B1A0D; padding: 8px 12px; border-radius: 4px; margin-top: 25px;"><b>⚠️ Mutable Elements Inside Tuples</b></h3>

<div style="border-left: 4px solid #FFB74D; padding-left: 15px; margin: 15px 0; color: #CCCCCC;">
    A tuple containing a list <code>t = ([1, 2], 3)</code> cannot be hashed (unhashable). The tuple itself is immutable, but the list inside can still be modified: <code>t[0].append(99)</code> works!
</div>

````python
import timeit

# Speed comparison
list_time = timeit.timeit('[1, 2, 3, 4, 5]', number=1000000)
tuple_time = timeit.timeit('(1, 2, 3, 4, 5)', number=1000000)
print(f'List creation:  {list_time:.3f}s')
print(f'Tuple creation: {tuple_time:.3f}s')
print(f'Tuple is {list_time/tuple_time:.1f}x faster')

# Tuples as dict keys
locations = {
    (40.7128, -74.0060): 'New York',
    (51.5074, -0.1278): 'London',
    (35.6762, 139.6503): 'Tokyo',
}
print(locations[(40.7128, -74.0060)])  # New York

# Mutable element inside tuple
t = ([1, 2], 'hello')
t[0].append(3)  # This works!
print(t)  # ([1, 2, 3], 'hello')
````

<h2 style="color: #000000ff; background-color: #02fff2ff; border-left: 6px solid #0c0c0cff; padding: 12px 15px; margin-top: 40px; border-radius: 4px;">
    <b>6. Professional Tuple Patterns : <span style="color: #070707ff; font-size: 0.85em; font-weight: normal;">Real-World Applications</span></b>
</h2>

<div style="border-left: 4px solid #4DB6AC; padding-left: 15px; margin: 15px 0; color: #CCCCCC;">
    <b>🔍 What is it?</b><br>
    Tuples shine in specific patterns: <b>multiple return values</b>, <b>dict items iteration</b>, <b>enumerate results</b>, and <b>zip pairings</b>. Recognizing where Python already uses tuples helps you write more idiomatic code.
</div>

**Where Python uses tuples implicitly:**

* `dict.items()` → `(key, value)` tuples
* `enumerate()` → `(index, value)` tuples
* `zip()` → tuples of paired elements
* Multiple return → `return a, b` packs a tuple
* String `%` formatting → `'%s is %d' % (name, age)`

<h3 style="color: #64B5F6; background-color: #0F1A2A; padding: 8px 12px; border-radius: 4px; margin-top: 25px;"><b>💼 Why Data Analysts Care</b></h3>

* **Multi-value returns**: `mean, std = get_statistics(data)` — clean function interfaces
* **Sorting with keys**: `sorted(data, key=lambda x: (x['dept'], -x['salary']))` — multi-level sort
* **Record comparison**: Tuples compare element-by-element: `(1, 'b') < (1, 'c')` → `True`

<h3 style="color: #81C784; background-color: #102414; padding: 8px 12px; border-radius: 4px; margin-top: 25px;"><b>🧠 Pro Tip</b></h3>

<div style="border-left: 4px solid #81C784; padding-left: 15px; margin: 15px 0; color: #CCCCCC;">
    Tuples compare <b>lexicographically</b>: element by element, left to right. Use this for multi-level sorting: <code>sorted(items, key=lambda x: (x.dept, -x.salary))</code>.
</div>

````python
# Multiple return values
def analyze(data):
    return min(data), max(data), sum(data)/len(data)

lo, hi, avg = analyze([10, 20, 30, 40, 50])
print(f'Min: {lo}, Max: {hi}, Avg: {avg}')

# dict.items() returns tuples
scores = {'Alice': 95, 'Bob': 82, 'Charlie': 91}
for name, score in scores.items():
    print(f'{name}: {score}')

# Multi-level sort with tuples
students = [('Bob', 85), ('Alice', 92), ('Bob', 78), ('Alice', 88)]
sorted_students = sorted(students)  # Sorts by name, then score
print(f'\nSorted: {sorted_students}')

# Tuple comparison
print((1, 'b') < (1, 'c'))  # True — compares second element
print((2, 'a') < (1, 'z'))  # False — first element decides
````
