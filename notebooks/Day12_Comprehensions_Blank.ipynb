<h2 style="color: #000000ff; background-color: #02fff2ff; border-left: 6px solid #0c0c0cff; padding: 12px 15px; margin-top: 40px; border-radius: 4px;">
    <b>1. Advanced List Comprehensions : <span style="color: #070707ff; font-size: 0.85em; font-weight: normal;">Deep Dive</span></b>
</h2>

<div style="border-left: 4px solid #4DB6AC; padding-left: 15px; margin: 15px 0; color: #CCCCCC;">
    <b>🔍 What is it?</b><br>
    We've seen basic list comprehensions. Advanced comprehensions allow for <b>nested loops</b> and <b>if/else conditionals</b>. They are the fastest, most Pythonic way to filter and transform lists.
</div>

```python
# If/Else mapping (Ternary inside comprehension)
results = ['Pass' if x >= 50 else 'Fail' for x in scores]

# Nested Loops (Flattening a matrix)
flat = [item for row in matrix for item in row]
```

<h3 style="color: #64B5F6; background-color: #0F1A2A; padding: 8px 12px; border-radius: 4px; margin-top: 25px;"><b>💼 Why Data Analysts Care</b></h3>

* **Data Cleaning**: Instantly mapping raw values to standardized categories
* **Matrix Flattening**: Converting a list-of-lists (JSON structure) into a flat 1D list

<h3 style="color: #FFB74D; background-color: #2B1A0D; padding: 8px 12px; border-radius: 4px; margin-top: 25px;"><b>⚠️ Readability Trap</b></h3>

<div style="border-left: 4px solid #FFB74D; padding-left: 15px; margin: 15px 0; color: #CCCCCC;">
    Never nest more than two loops in a comprehension. If it takes more than 5 seconds to read, convert it to standard <code>for</code> loops.
</div>

````python
# Nested loops in comprehensions
matrix = [[1, 2], [3, 4], [5, 6]]
flat = [num for row in matrix for num in row]
print(f'Flattened: {flat}')

# Conditionals (Filtering vs Mapping)
# 1. Filtering (if at the end)
evens = [x for x in flat if x % 2 == 0]
print(f'Evens: {evens}')

# 2. Mapping (if/else at the front)
labels = ['Even' if x % 2 == 0 else 'Odd' for x in flat]
print(f'Labels: {labels}')
````

<h2 style="color: #000000ff; background-color: #02fff2ff; border-left: 6px solid #0c0c0cff; padding: 12px 15px; margin-top: 40px; border-radius: 4px;">
    <b>2. Dict & Set Comprehensions : <span style="color: #070707ff; font-size: 0.85em; font-weight: normal;">Dynamic Hash Maps</span></b>
</h2>

<div style="border-left: 4px solid #4DB6AC; padding-left: 15px; margin: 15px 0; color: #CCCCCC;">
    <b>🔍 What is it?</b><br>
    Just like lists, you can create <b>Dictionaries</b> and <b>Sets</b> using comprehensions. For dictionaries, use the <code>key: value</code> syntax inside curly braces <code>{}</code>. For sets, just use values inside curly braces.
</div>

```python
# Dict Comprehension (Reverse mapping)
lookup = {value: key for key, value in original.items()}

# Set Comprehension (Unique lengths)
lengths = {len(word) for word in words}
```

<h3 style="color: #64B5F6; background-color: #0F1A2A; padding: 8px 12px; border-radius: 4px; margin-top: 25px;"><b>💼 Why Data Analysts Care</b></h3>

* **Fast Lookups**: Convert a list of user objects into a dictionary keyed by ID for O(1) lookups
* **Deduplication**: Extracting a set of unique domains from a list of email strings

<h3 style="color: #81C784; background-color: #102414; padding: 8px 12px; border-radius: 4px; margin-top: 25px;"><b>🧠 Pro Tip</b></h3>

<div style="border-left: 4px solid #81C784; padding-left: 15px; margin: 15px 0; color: #CCCCCC;">
    When swapping keys and values using <code>{v: k for k,v in d.items()}</code>, remember that if values are not unique, the last one processed will overwrite the others!
</div>

````python
names = ['Alice', 'Bob', 'Charlie', 'Bob']

# Set comprehension (Deduplicates automatically)
unique_lengths = {len(name) for name in names}
print(f'Unique lengths: {unique_lengths}')

# Dict comprehension
name_to_len = {name: len(name) for name in set(names)}
print(f'Dict map: {name_to_len}')

# Filtering a dictionary
prices = {'apple': 1.0, 'banana': 0.5, 'cherry': 2.0}
expensive = {k: v for k, v in prices.items() if v >= 1.0}
print(f'Expensive: {expensive}')
````
