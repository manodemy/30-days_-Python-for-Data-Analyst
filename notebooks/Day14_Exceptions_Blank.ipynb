<h2 style="color: #000000ff; background-color: #02fff2ff; border-left: 6px solid #0c0c0cff; padding: 12px 15px; margin-top: 40px; border-radius: 4px;">
    <b>1. Try / Except Blocks : <span style="color: #070707ff; font-size: 0.85em; font-weight: normal;">Catching Errors</span></b>
</h2>

<div style="border-left: 4px solid #4DB6AC; padding-left: 15px; margin: 15px 0; color: #CCCCCC;">
    <b>🔍 What is it?</b><br>
    Code crashes when an error occurs. A <code>try/except</code> block allows you to 'catch' exceptions, handle them gracefully, and let the program continue running. This is critical for data pipelines where one bad row shouldn't stop the whole job.
</div>

```python
try:
    result = 10 / 0
except ZeroDivisionError:
    result = 0
```

<h3 style="color: #64B5F6; background-color: #0F1A2A; padding: 8px 12px; border-radius: 4px; margin-top: 25px;"><b>💼 Why Data Analysts Care</b></h3>

* **API Calls**: Handling network timeouts or 500 errors without crashing
* **Data Cleaning**: Catching `ValueError` when trying to cast the string 'N/A' to an integer

<h3 style="color: #FFB74D; background-color: #2B1A0D; padding: 8px 12px; border-radius: 4px; margin-top: 25px;"><b>⚠️ The Bare Except Trap</b></h3>

<div style="border-left: 4px solid #FFB74D; padding-left: 15px; margin: 15px 0; color: #CCCCCC;">
    Never write a bare <code>except:</code>. It will catch SystemExit and KeyboardInterrupt (Ctrl+C), making your program impossible to stop. ALWAYS specify the exception type: <code>except Exception as e:</code> at a minimum.
</div>

````python
data = ['10', '20', 'invalid', '30']
clean_data = []

for item in data:
    try:
        # Try to convert to integer
        val = int(item)
        clean_data.append(val)
    except ValueError:
        # If it fails, log it and continue
        print(f"Warning: '{item}' is not a valid number. Skipping.")

print(f'Clean data: {clean_data}')
````

<h2 style="color: #000000ff; background-color: #02fff2ff; border-left: 6px solid #0c0c0cff; padding: 12px 15px; margin-top: 40px; border-radius: 4px;">
    <b>2. Else and Finally : <span style="color: #070707ff; font-size: 0.85em; font-weight: normal;">Complete Error Control</span></b>
</h2>

<div style="border-left: 4px solid #4DB6AC; padding-left: 15px; margin: 15px 0; color: #CCCCCC;">
    <b>🔍 What is it?</b><br>
    The full exception block includes <code>else</code> and <code>finally</code>. <code>else</code> runs <b>only if no exception occurred</b>. <code>finally</code> runs <b>no matter what</b> (even if the program crashes or returns early).
</div>

| Block | When does it run? | Common Use |
| :--- | :--- | :--- |
| `try` | Always | Risky code |
| `except` | If error happens | Error handling & logging |
| `else` | If NO error happens | Code that depends on the try block succeeding |
| `finally` | ALWAYS | Cleaning up resources (closing files/connections) |

<h3 style="color: #64B5F6; background-color: #0F1A2A; padding: 8px 12px; border-radius: 4px; margin-top: 25px;"><b>💼 Why Data Analysts Care</b></h3>

* **Database Connections**: Ensuring `db.close()` runs in the `finally` block even if the query fails
* **Safe Transactions**: Committing a database transaction in the `else` block only if the execution succeeds

<h3 style="color: #81C784; background-color: #102414; padding: 8px 12px; border-radius: 4px; margin-top: 25px;"><b>🧠 Pro Tip</b></h3>

<div style="border-left: 4px solid #81C784; padding-left: 15px; margin: 15px 0; color: #CCCCCC;">
    Keep the <code>try</code> block as small as possible. Put the risky code in `try`, and the code that depends on it succeeding in the `else` block.
</div>

````python
def divide(a, b):
    try:
        result = a / b
    except ZeroDivisionError:
        print("Error: Division by zero")
    else:
        # Runs only if try succeeded
        print(f"Success! Result is {result}")
    finally:
        # Runs no matter what
        print("Execution finished.\n")

divide(10, 2)
divide(10, 0)
````

<h2 style="color: #000000ff; background-color: #02fff2ff; border-left: 6px solid #0c0c0cff; padding: 12px 15px; margin-top: 40px; border-radius: 4px;">
    <b>3. Raising Exceptions : <span style="color: #070707ff; font-size: 0.85em; font-weight: normal;">Signaling Invalid State</span></b>
</h2>

<div style="border-left: 4px solid #4DB6AC; padding-left: 15px; margin: 15px 0; color: #CCCCCC;">
    <b>🔍 What is it?</b><br>
    Sometimes your code detects an invalid state (like a negative age). You can <b>manually trigger an error</b> using the <code>raise</code> keyword. This tells the calling function that something went wrong.
</div>

```python
def set_age(age):
    if age < 0:
        raise ValueError("Age cannot be negative")
    return age
```

<h3 style="color: #64B5F6; background-color: #0F1A2A; padding: 8px 12px; border-radius: 4px; margin-top: 25px;"><b>💼 Why Data Analysts Care</b></h3>

* **Data Validation**: Rejecting dirty data explicitly before it poisons the database
* **API Contracts**: Raising 400 Bad Request errors when user input is invalid

<h3 style="color: #81C784; background-color: #102414; padding: 8px 12px; border-radius: 4px; margin-top: 25px;"><b>🧠 Pro Tip</b></h3>

<div style="border-left: 4px solid #81C784; padding-left: 15px; margin: 15px 0; color: #CCCCCC;">
    You can define custom exceptions by subclassing `Exception`: <code>class DataValidationError(Exception): pass</code>. This makes your error handling highly specific.
</div>

````python
class InvalidTransactionError(Exception):
    """Custom exception for banking errors."""
    pass

def process_payment(amount):
    if amount <= 0:
        # Raise a built-in error
        raise ValueError("Payment amount must be positive")
    if amount > 10000:
        # Raise our custom error
        raise InvalidTransactionError("Transaction too large. Requires approval.")
    print(f"Processed ${amount}")

try:
    process_payment(50000)
except InvalidTransactionError as e:
    print(f"Blocked: {e}")
````
