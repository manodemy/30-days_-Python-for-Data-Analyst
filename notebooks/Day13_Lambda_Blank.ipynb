<h2 style="color: #000000ff; background-color: #02fff2ff; border-left: 6px solid #0c0c0cff; padding: 12px 15px; margin-top: 40px; border-radius: 4px;">
    <b>1. Lambda Functions : <span style="color: #070707ff; font-size: 0.85em; font-weight: normal;">Anonymous Inline Functions</span></b>
</h2>

<div style="border-left: 4px solid #4DB6AC; padding-left: 15px; margin: 15px 0; color: #CCCCCC;">
    <b>🔍 What is it?</b><br>
    A <b>lambda</b> is a small, anonymous, single-line function. Defined using <code>lambda arguments: expression</code>. It implicitly returns the result of the expression.
</div>

```python
# Standard
def add(x, y): return x + y
# Lambda
add_lambda = lambda x, y: x + y
```

<h3 style="color: #64B5F6; background-color: #0F1A2A; padding: 8px 12px; border-radius: 4px; margin-top: 25px;"><b>💼 Why Data Analysts Care</b></h3>

* **Pandas Apply**: `df['col'].apply(lambda x: x * 2)` — inline data transformations
* **Sorting**: `sorted(data, key=lambda d: d['age'])` — custom sort keys on the fly

<h3 style="color: #FFB74D; background-color: #2B1A0D; padding: 8px 12px; border-radius: 4px; margin-top: 25px;"><b>⚠️ Overusing Lambdas</b></h3>

<div style="border-left: 4px solid #FFB74D; padding-left: 15px; margin: 15px 0; color: #CCCCCC;">
    If a lambda gets complex or requires multiple lines, write a standard <code>def</code> function instead. Code readability is more important than brevity.
</div>

````python
# Sorting a list of tuples by the second element
records = [('Alice', 95), ('Bob', 82), ('Charlie', 91)]
sorted_records = sorted(records, key=lambda x: x[1])
print(sorted_records)

# Inline math mapping
math_funcs = {
    'double': lambda x: x * 2,
    'square': lambda x: x ** 2
}
print(math_funcs['square'](5))
````

<h2 style="color: #000000ff; background-color: #02fff2ff; border-left: 6px solid #0c0c0cff; padding: 12px 15px; margin-top: 40px; border-radius: 4px;">
    <b>2. Map, Filter, Reduce : <span style="color: #070707ff; font-size: 0.85em; font-weight: normal;">Functional Paradigms</span></b>
</h2>

<div style="border-left: 4px solid #4DB6AC; padding-left: 15px; margin: 15px 0; color: #CCCCCC;">
    <b>🔍 What is it?</b><br>
    Python supports functional programming concepts. <code>map()</code> applies a function to all items. <code>filter()</code> keeps items where a function returns True. <code>reduce()</code> (from `functools`) repeatedly applies a function to pairs of items to reduce a list to a single value.
</div>

| Function | Purpose | Equivalent Comprehension |
| :--- | :--- | :--- |
| `map(f, data)` | Transform | `[f(x) for x in data]` |
| `filter(f, data)`| Extract | `[x for x in data if f(x)]` |
| `reduce(f, data)`| Aggregate | (No direct equivalent) |

<h3 style="color: #64B5F6; background-color: #0F1A2A; padding: 8px 12px; border-radius: 4px; margin-top: 25px;"><b>💼 Why Data Analysts Care</b></h3>

* **Data Pipelines**: Chaining operations: `reduce(add, map(square, filter(is_even, data)))`
* **Legacy Code**: You will see `map/filter` frequently in codebases written by Java/JS developers

<h3 style="color: #81C784; background-color: #102414; padding: 8px 12px; border-radius: 4px; margin-top: 25px;"><b>🧠 Pro Tip</b></h3>

<div style="border-left: 4px solid #81C784; padding-left: 15px; margin: 15px 0; color: #CCCCCC;">
    In modern Python, list comprehensions are generally preferred over <code>map()</code> and <code>filter()</code> because they are more readable and slightly faster.
</div>

````python
from functools import reduce

nums = [1, 2, 3, 4, 5]

# Map (Transform)
doubled = list(map(lambda x: x * 2, nums))
print(f'Map: {doubled}')

# Filter (Extract)
evens = list(filter(lambda x: x % 2 == 0, nums))
print(f'Filter: {evens}')

# Reduce (Aggregate - e.g., multiply all)
product = reduce(lambda x, y: x * y, nums)
print(f'Reduce (Product): {product}')
````
