<h2 style="color: #000000ff; background-color: #02fff2ff; border-left: 6px solid #0c0c0cff; padding: 12px 15px; margin-top: 40px; border-radius: 4px;">
    <b>1. The GroupBy Mechanics : <span style="color: #070707ff; font-size: 0.85em; font-weight: normal;">Split-Apply-Combine</span></b>
</h2>

<div style="border-left: 4px solid #4DB6AC; padding-left: 15px; margin: 15px 0; color: #CCCCCC;">
    <b>🔍 What is it?</b><br>
    The <code>.groupby()</code> method is identical to the SQL <code>GROUP BY</code> statement. It follows the <b>Split-Apply-Combine</b> pattern: it <i>splits</i> the data into groups based on a key, <i>applies</i> an aggregation function (like sum or mean), and <i>combines</i> the results into a new DataFrame.
</div>

```python
# Calculate the average salary per department
# 1. Split by Dept | 2. Select Salary | 3. Apply mean()
avg_salary = df.groupby('Dept')['Salary'].mean()
```

<h3 style="color: #64B5F6; background-color: #0F1A2A; padding: 8px 12px; border-radius: 4px; margin-top: 25px;"><b>💼 Why Data Analysts Care</b></h3>

* **Financial Reporting**: Calculating total monthly revenue: `df.groupby('Month')['Revenue'].sum()`
* **User Behavior**: Finding the maximum session length per user: `df.groupby('UserID')['Duration'].max()`

<h3 style="color: #FFB74D; background-color: #2B1A0D; padding: 8px 12px; border-radius: 4px; margin-top: 25px;"><b>⚠️ Unaggregated GroupBys</b></h3>

<div style="border-left: 4px solid #FFB74D; padding-left: 15px; margin: 15px 0; color: #CCCCCC;">
    If you just print <code>df.groupby('Dept')</code>, you will get a `<DataFrameGroupBy object>` memory address. You MUST apply an aggregation function (like `.sum()`) to actually see data!
</div>

````python
import pandas as pd

df = pd.DataFrame({
    'Dept': ['IT', 'HR', 'IT', 'Sales', 'HR'],
    'Employee': ['Alice', 'Bob', 'Charlie', 'David', 'Eve'],
    'Salary': [70000, 50000, 80000, 90000, 60000],
    'Experience': [3, 2, 5, 10, 4]
})

# Group by Department and calculate Mean for all numeric columns
print("Mean per Dept:\n", df.groupby('Dept').mean(numeric_only=True))

# Group by Department, select ONLY Salary, and calculate Sum
print("\nTotal Salary per Dept:\n", df.groupby('Dept')['Salary'].sum())
````

<h2 style="color: #000000ff; background-color: #02fff2ff; border-left: 6px solid #0c0c0cff; padding: 12px 15px; margin-top: 40px; border-radius: 4px;">
    <b>2. Multiple Aggregations (.agg) : <span style="color: #070707ff; font-size: 0.85em; font-weight: normal;">Advanced Summaries</span></b>
</h2>

<div style="border-left: 4px solid #4DB6AC; padding-left: 15px; margin: 15px 0; color: #CCCCCC;">
    <b>🔍 What is it?</b><br>
    Sometimes you want multiple statistics at once (e.g., the Mean AND the Max). You can pass a list of functions to the <b><code>.agg()</code></b> method. You can also pass a dictionary to apply different functions to different columns!
</div>

```python
# Min and Max for Salary
df.groupby('Dept')['Salary'].agg(['min', 'max'])

# Different math for different columns
df.groupby('Dept').agg({'Salary': 'mean', 'Experience': 'max'})
```

<h3 style="color: #64B5F6; background-color: #0F1A2A; padding: 8px 12px; border-radius: 4px; margin-top: 25px;"><b>💼 Why Data Analysts Care</b></h3>

* **Executive Dashboards**: Showing the Total Sales, Average Order Value, and Count of Orders per region in one table

<h3 style="color: #81C784; background-color: #102414; padding: 8px 12px; border-radius: 4px; margin-top: 25px;"><b>🧠 Pro Tip</b></h3>

<div style="border-left: 4px solid #81C784; padding-left: 15px; margin: 15px 0; color: #CCCCCC;">
    GroupBy outputs have the group keys (e.g., 'Dept') as the <b>Index</b>. If you want a flat standard DataFrame, append <code>.reset_index()</code> to the end of your groupby chain!
</div>

````python
import pandas as pd

df = pd.DataFrame({
    'Dept': ['IT', 'HR', 'IT', 'Sales', 'HR'],
    'Salary': [70000, 50000, 80000, 90000, 60000],
    'Experience': [3, 2, 5, 10, 4]
})

# Multiple aggregations on a single column
stats = df.groupby('Dept')['Salary'].agg(['mean', 'sum', 'count'])
print("Salary Stats:\n", stats)

# Dictionary aggregation (Different rules per column)
complex_agg = df.groupby('Dept').agg({
    'Salary': 'sum',       # Total budget
    'Experience': 'max'    # Most senior person
}).reset_index()           # Flattens the index!

print("\nComplex Aggregation:\n", complex_agg)
````

<h2 style="color: #000000ff; background-color: #02fff2ff; border-left: 6px solid #0c0c0cff; padding: 12px 15px; margin-top: 40px; border-radius: 4px;">
    <b>3. Transform & Apply : <span style="color: #070707ff; font-size: 0.85em; font-weight: normal;">Row-Level Group Math</span></b>
</h2>

<div style="border-left: 4px solid #4DB6AC; padding-left: 15px; margin: 15px 0; color: #CCCCCC;">
    <b>🔍 What is it?</b><br>
    While `.agg()` collapses data into a summary table, <b><code>.transform()</code></b> returns data the exact same shape as the original. This is used for calculating group-level metrics and broadcasting them back to the original rows (like SQL Window Functions).
</div>

```python
# Calculate the Dept Mean, and broadcast it to every employee's row
df['Dept_Avg'] = df.groupby('Dept')['Salary'].transform('mean')

# Now you can calculate how much above/below average they are!
df['Above_Avg'] = df['Salary'] - df['Dept_Avg']
```

<h3 style="color: #64B5F6; background-color: #0F1A2A; padding: 8px 12px; border-radius: 4px; margin-top: 25px;"><b>💼 Why Data Analysts Care</b></h3>

* **Standardization**: Z-Score normalization per category: `(x - group_mean) / group_std`
* **Percent to Total**: Calculating what percentage an employee's salary contributes to their department's total budget

<h3 style="color: #81C784; background-color: #102414; padding: 8px 12px; border-radius: 4px; margin-top: 25px;"><b>🧠 Pro Tip</b></h3>

<div style="border-left: 4px solid #81C784; padding-left: 15px; margin: 15px 0; color: #CCCCCC;">
    Use <code>.agg()</code> when you want to reduce dimensions (create a summary report). Use <code>.transform()</code> when you want to add new calculated columns to your existing DataFrame.
</div>

````python
import pandas as pd

df = pd.DataFrame({
    'Dept': ['IT', 'IT', 'HR', 'HR'],
    'Name': ['Alice', 'Bob', 'Charlie', 'David'],
    'Salary': [80000, 60000, 50000, 70000]
})

# Calculate the mean per department and append it as a new column
df['Dept_Mean'] = df.groupby('Dept')['Salary'].transform('mean')

# Calculate the difference
df['Diff_From_Mean'] = df['Salary'] - df['Dept_Mean']

print("Transformed DataFrame:\n", df)
````
