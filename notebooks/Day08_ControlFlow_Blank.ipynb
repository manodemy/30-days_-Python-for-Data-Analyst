</h2>

<div style="border-left: 4px solid #4DB6AC; padding-left: 15px; margin: 15px 0; color: #CCCCCC;">
    <b>🔍 What is it?</b><br>
    The <code>if</code> statement evaluates an expression. If it is <b>truthy</b>, the indented block runs. </div>

```python
if condition1:
    # runs if condition1 is True
elif condition2:
    # runs if condition1 is False AND condition2 is True
else:
    # runs if all above are False
```

<h3 style="color: #64B5F6; background-color: #0F1A2A; padding: 8px 12px; border-radius: 4px; margin-top: 25px;"><b>💼 Why Data Analysts Care</b></h3>

* **Null Checking**: `if data:` — checking if a list/dict is not empty

<div style="border-left: 4px solid #FFB74D; padding-left: 15px; margin: 15px 0; color: #CCCCCC;">
    Do not write <code>if len(my_list) > 0:</code>. It is unpythonic and slower. Just write <code>if my_list:</code> because empty lists are falsy.
</div>

````python
# Truthy / Falsy evaluation
empty_list = []
name = "Alice"
zero = 0

if empty_list:
    print('List has data')
else:
    print('List is empty (falsy)')

if name:
    print(f'Name is {name}')

# Classic if/elif/else routing
status_code = 404

if status_code == 200:
    print('OK')
elif status_code == 404:
    print('Not Found')
elif status_code >= 500:
    print('Server Error')
else:
    print('Unknown Status')
````

<h2 style="color: #000000ff; background-color: #02fff2ff; border-left: 6px solid #0c0c0cff; padding: 12px 15px; margin-top: 40px; border-radius: 4px;">
    <b>2. Ternary Operator : <span style="color: #070707ff; font-size: 0.85em; font-weight: normal;">Inline Conditionals</span></b>
</h2>

<div style="border-left: 4px solid #4DB6AC; padding-left: 15px; margin: 15px 0; color: #CCCCCC;">
    <b>🔍 What is it?</b><br>
    The ternary operator <code>x if condition else y</code> evaluates to <code>x</code> if the condition is True, otherwise <code>y</code>. It is an <b>expression</b>, meaning it returns a value and can be assigned to a variable or used inline.
</div>

| Approach | Code | Lines |
| :--- | :--- | :--- |
| Standard | `if x > 0: y = 1`<br>`else: y = 0` | 4 |
| Ternary | `y = 1 if x > 0 else 0` | 1 |

<h3 style="color: #64B5F6; background-color: #0F1A2A; padding: 8px 12px; border-radius: 4px; margin-top: 25px;"><b>💼 Why Data Analysts Care</b></h3>

* **Variable Assignment**: `status = 'active' if is_paid else 'inactive'`
* **Data Cleaning**: `val = 0 if val is None else val` — handle nulls
* **List Comprehensions**: `[x if x > 0 else 0 for x in data]` — inline mapping

<h3 style="color: #FFB74D; background-color: #2B1A0D; padding: 8px 12px; border-radius: 4px; margin-top: 25px;"><b>⚠️ Nested Ternaries</b></h3>

<div style="border-left: 4px solid #FFB74D; padding-left: 15px; margin: 15px 0; color: #CCCCCC;">
    Avoid nesting ternaries (e.g., <code>a if cond1 else b if cond2 else c</code>). It destroys readability. Use a standard `if/elif` block instead.
</div>

````python
age = 20

# Standard if/else
if age >= 18:
    category = 'Adult'
else:
    category = 'Minor'

# Ternary equivalent
category_ternary = 'Adult' if age >= 18 else 'Minor'
print(f'Category: {category_ternary}')

# Useful in f-strings
count = 1
print(f'I have {count} apple{"s" if count != 1 else ""}')

# Handling nulls
user_input = None
clean_input = user_input if user_input is not None else 'default_value'
print(f'Cleaned: {clean_input}')
````

<h2 style="color: #000000ff; background-color: #02fff2ff; border-left: 6px solid #0c0c0cff; padding: 12px 15px; margin-top: 40px; border-radius: 4px;">
    <b>3. Match-Case (Python 3.10+) : <span style="color: #070707ff; font-size: 0.85em; font-weight: normal;">Structural Pattern Matching</span></b>
</h2>

<div style="border-left: 4px solid #4DB6AC; padding-left: 15px; margin: 15px 0; color: #CCCCCC;">
    <b>🔍 What is it?</b><br>
    Introduced in Python 3.10, <code>match-case</code> is much more than a switch statement. It allows you to <b>match values AND structural patterns</b> (like the shape of a dictionary or list) and extract variables simultaneously.
</div>

```python
match status_code:
    case 200 | 201:
        print('Success')
    case 404:
        print('Not Found')
    case _:
        print('Unknown/Default')
```

<h3 style="color: #64B5F6; background-color: #0F1A2A; padding: 8px 12px; border-radius: 4px; margin-top: 25px;"><b>💼 Why Data Analysts Care</b></h3>

* **API Routing**: Handle different types of incoming JSON payloads based on their structure
* **Command Parsing**: Parse user input like `['drop', 'table', name]` cleanly
* **State Machines**: Cleanly manage transitions between app states

<h3 style="color: #FFB74D; background-color: #2B1A0D; padding: 8px 12px; border-radius: 4px; margin-top: 25px;"><b>⚠️ Not Just a Switch</b></h3>

<div style="border-left: 4px solid #FFB74D; padding-left: 15px; margin: 15px 0; color: #CCCCCC;">
    If you are just comparing a single variable against constants (like a switch), a dictionary lookup is often faster and cleaner: <code>routes.get(status, 'Unknown')</code>.
</div>

````python
# Simple literal matching
command = 'start'

match command:
    case 'start' | 'run':  # Use | for multiple values
        print('System starting...')
    case 'stop':
        print('System stopping...')
    case _:  # The underscore acts as the default/else case
        print('Unknown command')

# Matching with variable capture
user_input = ['greet', 'Alice']

match user_input:
    case ['greet', name]:  # Matches a 2-element list starting with 'greet'
        print(f'Hello, {name}!')
    case ['exit']:
        print('Goodbye!')
    case _:
        print('Invalid syntax')
````

    <b>4. Pattern Matching with Dicts : <span style="color: #070707ff; font-size: 0.85em; font-weight: normal;">Extracting JSON Payloads</span></b>
</h2>

<div style="border-left: 4px solid #4DB6AC; padding-left: 15px; margin: 15px 0; color: #CCCCCC;">
    <b>🔍 What is it?</b><br>
    The true power of <code>match-case</code> lies in matching dictionaries and extracting specific keys. It ignores extra keys, making it perfect for parsing heterogeneous JSON payloads.
</div>

```python
payload = {'type': 'user_update', 'id': 101, 'name': 'Alice'}

match payload:
    case {'type': 'user_update', 'id': uid}:
        print(f'Updating user {uid}')
```

<h3 style="color: #64B5F6; background-color: #0F1A2A; padding: 8px 12px; border-radius: 4px; margin-top: 25px;"><b>💼 Why Data Analysts Care</b></h3>

* **Event Streams**: Process streams of JSON events (like Kafka topics) where schemas vary
* **Data Validation**: Ensure a dictionary has a specific shape before processing
* **AST Parsing**: Traversing Abstract Syntax Trees or complex nested data

<h3 style="color: #81C784; background-color: #102414; padding: 8px 12px; border-radius: 4px; margin-top: 25px;"><b>🧠 Pro Tip</b></h3>

<div style="border-left: 4px solid #81C784; padding-left: 15px; margin: 15px 0; color: #CCCCCC;">
    In dictionary matching, extra keys are ignored. <code>case {'id': uid}:</code> matches <i>any</i> dictionary that has an 'id' key, regardless of what else is in it.
</div>

````python
events = [
    {'type': 'click', 'x': 10, 'y': 20},
    {'type': 'keypress', 'key': 'Enter'},
    {'type': 'unknown', 'data': '???', 'id': 99}
]

for event in events:
    match event:
        # Match dictionary shape and capture variables
        case {'type': 'click', 'x': x_pos, 'y': y_pos}:
            print(f'Mouse click at ({x_pos}, {y_pos})')
            
        case {'type': 'keypress', 'key': key_name}:
            print(f'Keyboard hit: {key_name}')
            
        # Use **rest to capture the remaining keys
        case {'type': unknown_type, **rest}:
            print(f'Unhandled event: {unknown_type}, payload: {rest}')
````

<h2 style="color: #000000ff; background-color: #02fff2ff; border-left: 6px solid #0c0c0cff; padding: 12px 15px; margin-top: 40px; border-radius: 4px;">
    <b>5. Guard Clauses & Early Returns : <span style="color: #070707ff; font-size: 0.85em; font-weight: normal;">Professional Logic Design</span></b>
</h2>

<div style="border-left: 4px solid #4DB6AC; padding-left: 15px; margin: 15px 0; color: #CCCCCC;">
    <b>🔍 What is it?</b><br>
    A <b>guard clause</b> is an early <code>return</code> (or <code>continue</code>/<code>break</code>) that handles an edge case at the top of a function. This prevents deep nesting (the 'Arrow Anti-Pattern') and keeps the 'happy path' unindented at the bottom.
</div>

```python
# ❌ Bad: Nested 'Arrow' code
def process(data):
    if data is not None:
        if len(data) > 0:
            # Do actual work here
            return True
    return False

# ✅ Good: Guard Clauses
def process(data):
    if not data:           # Guard 1
        return False
    if len(data) == 0:     # Guard 2
        return False
    
    # Happy path (unindented!)
    return True
```

<h3 style="color: #64B5F6; background-color: #0F1A2A; padding: 8px 12px; border-radius: 4px; margin-top: 25px;"><b>💼 Why Data Analysts Care</b></h3>

* **Code Review**: Flat code is infinitely easier to read than deeply nested if-statements
* **Data Validation**: Validating DataFrame structures at the top of a pipeline step
* **Error Handling**: Failing fast saves computation time

<h3 style="color: #81C784; background-color: #102414; padding: 8px 12px; border-radius: 4px; margin-top: 25px;"><b>🧠 Pro Tip</b></h3>

<div style="border-left: 4px solid #81C784; padding-left: 15px; margin: 15px 0; color: #CCCCCC;">
    Rule of thumb: <b>Fail fast, return early.</b> Handle all your errors, nulls, and edge cases in the first few lines of your function.
</div>

````python
def calculate_discount(user, cart_value):
    # Guard 1: Invalid input
    if not user or cart_value <= 0:
        return 0.0
        
    # Guard 2: Ineligible tier
    if user.get('tier') == 'free':
        return 0.0
        
    # Happy Path: Actual business logic
    if user.get('tier') == 'premium':
        return cart_value * 0.20
    
    return cart_value * 0.05

# Clean, readable execution
print(calculate_discount({'tier': 'free'}, 100))     # 0.0
print(calculate_discount({'tier': 'premium'}, 100))  # 20.0
````

<h2 style="color: #000000ff; background-color: #02fff2ff; border-left: 6px solid #0c0c0cff; padding: 12px 15px; margin-top: 40px; border-radius: 4px;">
    <b>6. Complex Boolean Logic : <span style="color: #070707ff; font-size: 0.85em; font-weight: normal;">De Morgan's Laws</span></b>
</h2>

<div style="border-left: 4px solid #4DB6AC; padding-left: 15px; margin: 15px 0; color: #CCCCCC;">
    <b>🔍 What is it?</b><br>
    Writing complex <code>if</code> statements often leads to logic bugs. <b>De Morgan's Laws</b> are mathematical rules for simplifying negated boolean expressions. They are essential for refactoring messy conditionals.
</div>

**De Morgan's Laws:**

1. `not (A and B)` ≡ `(not A) or (not B)`
2. `not (A or B)`  ≡ `(not A) and (not B)`

```python
# ❌ Hard to read:
if not (age >= 18 and status == 'active'):

# ✅ De Morgan equivalent:
if age < 18 or status != 'active':
```

<h3 style="color: #64B5F6; background-color: #0F1A2A; padding: 8px 12px; border-radius: 4px; margin-top: 25px;"><b>💼 Why Data Analysts Care</b></h3>

* **Filter Simplification**: Refactoring complex pandas queries: `~((df.A > 0) & (df.B == 1))` into `(df.A <= 0) | (df.B != 1)`
* **Bug Prevention**: Negated `and`/`or` chains are the #1 source of logic errors in code reviews

<h3 style="color: #FFB74D; background-color: #2B1A0D; padding: 8px 12px; border-radius: 4px; margin-top: 25px;"><b>⚠️ Excessive Negation</b></h3>

<div style="border-left: 4px solid #FFB74D; padding-left: 15px; margin: 15px 0; color: #CCCCCC;">
    Avoid double negatives like <code>if not is_invalid:</code>. Name your variables positively: <code>if is_valid:</code>.
</div>

````python
age = 16
has_permission = True

# Example 1: not (A and B) == not A or not B
# "You cannot enter if you are not an adult and you don't have permission"

# Hard to parse:
if not (age >= 18 and has_permission):
    print("Access Denied (v1)")

# Easier to parse (De Morgan):
if age < 18 or not has_permission:
    print("Access Denied (v2)")

# Example 2: not (A or B) == not A and not B
is_weekend = False
is_holiday = False

if not (is_weekend or is_holiday):
    print("It's a workday")

if not is_weekend and not is_holiday:
    print("It's a workday (cleaner)")
````

<h2 style="color: #000000ff; background-color: #02fff2ff; border-left: 6px solid #0c0c0cff; padding: 12px 15px; margin-top: 40px; border-radius: 4px;">
    <b>7. Exceptions as Control Flow : <span style="color: #070707ff; font-size: 0.85em; font-weight: normal;">EAFP vs LBYL</span></b>
</h2>

<div style="border-left: 4px solid #4DB6AC; padding-left: 15px; margin: 15px 0; color: #CCCCCC;">
    <b>🔍 What is it?</b><br>
    Python embraces the <b>EAFP</b> philosophy: <i>Easier to Ask for Forgiveness than Permission</i>. Instead of checking if something is valid before doing it (<b>LBYL</b>: <i>Look Before You Leap</i>), just try it and catch the exception. It is often faster and cleaner.
</div>

| Approach | Philosophy | Code Example |
| :--- | :--- | :--- |
| LBYL | Look Before You Leap | `if key in d: val = d[key]` |
| EAFP | Ask Forgiveness | `try: val = d[key]`<br>`except KeyError: pass` |

<h3 style="color: #64B5F6; background-color: #0F1A2A; padding: 8px 12px; border-radius: 4px; margin-top: 25px;"><b>💼 Why Data Analysts Care</b></h3>

* **Data Parsing**: Trying to cast strings to integers: `try: int(val) except ValueError: val = 0`
* **File Operations**: Trying to open a file instead of checking if it exists first (which causes race conditions)

<h3 style="color: #FFB74D; background-color: #2B1A0D; padding: 8px 12px; border-radius: 4px; margin-top: 25px;"><b>⚠️ Naked Excepts</b></h3>

<div style="border-left: 4px solid #FFB74D; padding-left: 15px; margin: 15px 0; color: #CCCCCC;">
    Never write a bare <code>except:</code>. It catches system-exiting exceptions like <code>KeyboardInterrupt</code> (Ctrl+C). Always specify the exact exception: <code>except ValueError:</code>.
</div>

````python
# LBYL: Look Before You Leap (checking types/keys first)
data = {'age': '25'}
if 'age' in data and data['age'].isdigit():
    age = int(data['age'])
    print(f'LBYL Age: {age}')

# EAFP: Easier to Ask Forgiveness (Pythonic)
try:
    # Just try the operation!
    age2 = int(data['age'])
    print(f'EAFP Age: {age2}')
except (KeyError, ValueError):
    # Handle the specific failures
    print('Invalid or missing age')

# EAFP avoids race conditions when dealing with files
````
