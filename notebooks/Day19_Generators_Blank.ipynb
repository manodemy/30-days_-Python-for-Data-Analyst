<h2 style="color: #000000ff; background-color: #02fff2ff; border-left: 6px solid #0c0c0cff; padding: 12px 15px; margin-top: 40px; border-radius: 4px;">
    <b>1. Advanced Generators : <span style="color: #070707ff; font-size: 0.85em; font-weight: normal;">yield from & State</span></b>
</h2>

<div style="border-left: 4px solid #4DB6AC; padding-left: 15px; margin: 15px 0; color: #CCCCCC;">
    <b>🔍 What is it?</b><br>
    We've learned that <code>yield</code> pauses a function. Advanced generators can yield from OTHER generators using <b><code>yield from</code></b>, allowing you to chain pipelines cleanly. Generators also maintain complex internal state across iterations.
</div>

```python
def sub_gen():
    yield 1
    yield 2

def main_gen():
    yield from sub_gen() # Chains the sub-generator seamlessly
    yield 3
```

<h3 style="color: #64B5F6; background-color: #0F1A2A; padding: 8px 12px; border-radius: 4px; margin-top: 25px;"><b>💼 Why Data Analysts Care</b></h3>

* **Tree Traversal**: Flattening complex nested JSON structures recursively
* **Data Pipelining**: Chaining multiple generator functions (extract -> transform -> load)

<h3 style="color: #FFB74D; background-color: #2B1A0D; padding: 8px 12px; border-radius: 4px; margin-top: 25px;"><b>⚠️ Infinite Recursion</b></h3>

<div style="border-left: 4px solid #FFB74D; padding-left: 15px; margin: 15px 0; color: #CCCCCC;">
    When using <code>yield from</code> recursively, ensure you have a base case that stops yielding, just like standard recursion, to prevent a RecursionError.
</div>

````python
def flatten_list(nested):
    """Recursively flatten a list of lists using yield from"""
    for item in nested:
        if isinstance(item, list):
            yield from flatten_list(item)  # Delegate to sub-generator
        else:
            yield item

data = [1, [2, 3], [4, [5, 6]]]
flat = list(flatten_list(data))
print(f"Original: {data}")
print(f"Flattened: {flat}")
````

<h2 style="color: #000000ff; background-color: #02fff2ff; border-left: 6px solid #0c0c0cff; padding: 12px 15px; margin-top: 40px; border-radius: 4px;">
    <b>2. Coroutines : <span style="color: #070707ff; font-size: 0.85em; font-weight: normal;">Sending Data to Generators</span></b>
</h2>

<div style="border-left: 4px solid #4DB6AC; padding-left: 15px; margin: 15px 0; color: #CCCCCC;">
    <b>🔍 What is it?</b><br>
    Generators don't just produce data; they can consume it! By using the <b><code>.send()</code></b> method, you can pass data <i>into</i> a paused generator. A generator used this way is called a <b>Coroutine</b>.
</div>

```python
def accumulator():
    total = 0
    while True:
        value = yield total # Pauses here, waits for .send()
        total += value
```

<h3 style="color: #64B5F6; background-color: #0F1A2A; padding: 8px 12px; border-radius: 4px; margin-top: 25px;"><b>💼 Why Data Analysts Care</b></h3>

* **Event Loops**: Coroutines are the historical foundation of Python's `asyncio` for asynchronous programming
* **State Machines**: Building state machines that wait for external inputs to transition states

<h3 style="color: #81C784; background-color: #102414; padding: 8px 12px; border-radius: 4px; margin-top: 25px;"><b>🧠 Pro Tip</b></h3>

<div style="border-left: 4px solid #81C784; padding-left: 15px; margin: 15px 0; color: #CCCCCC;">
    Before you can <code>.send()</code> a value into a coroutine, you MUST 'prime' it by calling <code>next(gen)</code> or <code>gen.send(None)</code> so it reaches the first <code>yield</code> statement.
</div>

````python
def averager():
    """Coroutine that yields the running average"""
    total = 0.0
    count = 0
    average = None
    while True:
        # Yield the current average, and WAIT for a new term
        term = yield average
        if term is None:
            break
        total += term
        count += 1
        average = total / count

# Usage
coro = averager()
next(coro)  # Prime the coroutine!

print(f"Sent 10, Avg: {coro.send(10)}")
print(f"Sent 20, Avg: {coro.send(20)}")
print(f"Sent 30, Avg: {coro.send(30)}")
````

<h2 style="color: #000000ff; background-color: #02fff2ff; border-left: 6px solid #0c0c0cff; padding: 12px 15px; margin-top: 40px; border-radius: 4px;">
    <b>3. Context Managers : <span style="color: #070707ff; font-size: 0.85em; font-weight: normal;">Building Your Own 'with'</span></b>
</h2>

<div style="border-left: 4px solid #4DB6AC; padding-left: 15px; margin: 15px 0; color: #CCCCCC;">
    <b>🔍 What is it?</b><br>
    You know how to use <code>with open()</code>. You can build your own Context Managers using the <code>contextlib</code> module and a generator. This is the cleanest way to handle setup and teardown logic.
</div>

```python
from contextlib import contextmanager

@contextmanager
def my_timer():
    start = time.time()
    yield # Code inside the 'with' block runs here
    print(f'Took {time.time() - start}s')
```

<h3 style="color: #64B5F6; background-color: #0F1A2A; padding: 8px 12px; border-radius: 4px; margin-top: 25px;"><b>💼 Why Data Analysts Care</b></h3>

* **Database Connections**: Yielding a DB connection and ensuring `conn.close()` happens after the `with` block
* **Temporary State**: Temporarily changing a directory, then switching back safely

<h3 style="color: #81C784; background-color: #102414; padding: 8px 12px; border-radius: 4px; margin-top: 25px;"><b>🧠 Pro Tip</b></h3>

<div style="border-left: 4px solid #81C784; padding-left: 15px; margin: 15px 0; color: #CCCCCC;">
    If an exception occurs inside the <code>with</code> block, it will be raised at the <code>yield</code> statement in your context manager. Wrap the <code>yield</code> in a <code>try/finally</code> to ensure cleanup!
</div>

````python
from contextlib import contextmanager
import time

@contextmanager
def timing(label):
    """A context manager to measure execution time"""
    start = time.time()
    try:
        yield  # The indented code runs here!
    finally:
        # This runs when the block finishes (or crashes!)
        end = time.time()
        print(f"{label} took {end - start:.4f} seconds")

# Usage
with timing("Loop Test"):
    # Simulate work
    for _ in range(1_000_000):
        pass
    print("Work done!")
````
