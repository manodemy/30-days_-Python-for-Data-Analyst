</h2>

<div style="border-left: 4px solid #4DB6AC; padding-left: 15px; margin: 15px 0; color: #CCCCCC;">
    <b>🔍 What is it?</b><br>
    Welcome to Phase 2: Data Analytics. <b>NumPy</b> (Numerical Python) is the foundation of Pandas and Scikit-Learn. Its core structure is the <b>ndarray</b> (N-dimensional array), which is up to 50x faster than Python lists because it uses fixed-type C-arrays under the hood.

```python
import numpy as np

```

<h3 style="color: #64B5F6; background-color: #0F1A2A; padding: 8px 12px; border-radius: 4px; margin-top: 25px;"><b>💼 Why Data Analysts Care</b></h3>

* **Performance**: Processing millions of data points instantly
* **Memory Efficiency**: NumPy arrays take up significantly less RAM than standard Python lists

<h3 style="color: #FFB74D; background-color: #2B1A0D; padding: 8px 12px; border-radius: 4px; margin-top: 25px;"><b>⚠️ Mixed Data Types</b></h3>

<div style="border-left: 4px solid #FFB74D; padding-left: 15px; margin: 15px 0; color: #CCCCCC;">
    Unlike Python lists which can hold `[1, 'a', True]`, a NumPy array must be <b>homogenous</b> (all elements must be the same type, e.g., all floats or all ints). If you mix them, NumPy will force them into strings.
</div>

````python
import numpy as np

# Creating arrays
list_data = [1, 2, 3]
array_1d = np.array(list_data)
print(f"1D Array: {array_1d}")
print(f"Type: {type(array_1d)}")

# Creating 2D arrays (Matrices)
matrix = np.array([[1, 2], [3, 4]])
print(f"\n2D Array (Matrix):\n{matrix}")

# Checking attributes
print(f"\nShape: {matrix.shape}")   # (Rows, Cols)
print(f"Dimensions: {matrix.ndim}") # 2
print(f"Data Type: {matrix.dtype}") # int64 or int32
````

<h2 style="color: #000000ff; background-color: #02fff2ff; border-left: 6px solid #0c0c0cff; padding: 12px 15px; margin-top: 40px; border-radius: 4px;">
    <b>2. Array Generation : <span style="color: #070707ff; font-size: 0.85em; font-weight: normal;">Built-in Constructors</span></b>
</h2>

<div style="border-left: 4px solid #4DB6AC; padding-left: 15px; margin: 15px 0; color: #CCCCCC;">
    <b>🔍 What is it?</b><br>
    You rarely build massive arrays by manually typing lists. NumPy provides powerful generation functions like <code>np.zeros()</code>, <code>np.arange()</code>, and <code>np.linspace()</code>.
</div>

| Function | Purpose | Example |
| :--- | :--- | :--- |
| `np.zeros(shape)` | Array of 0s | `np.zeros((3,3))` |
| `np.ones(shape)` | Array of 1s | `np.ones(5)` |
| `np.arange(start, stop, step)`| Like Python `range` | `np.arange(0, 10, 2)` |
| `np.linspace(start, stop, num)`| Evenly spaced points | `np.linspace(0, 1, 5)` |

<h3 style="color: #64B5F6; background-color: #0F1A2A; padding: 8px 12px; border-radius: 4px; margin-top: 25px;"><b>💼 Why Data Analysts Care</b></h3>

* **Initialization**: Creating empty matrices to fill with model weights
* **Plotting**: Using `linspace` to generate perfectly spaced X-axis values for charts

<h3 style="color: #81C784; background-color: #102414; padding: 8px 12px; border-radius: 4px; margin-top: 25px;"><b>🧠 Pro Tip</b></h3>

<div style="border-left: 4px solid #81C784; padding-left: 15px; margin: 15px 0; color: #CCCCCC;">
    <code>np.arange()</code> works with floats (e.g., step=0.5), unlike Python's built-in <code>range()</code> which only takes integers!
</div>

````python
import numpy as np

# Zeros and Ones
zeros = np.zeros((2, 4))  # Pass shape as a tuple
print("Zeros:\n", zeros)

# Arange (Start, Stop exclusive, Step)
steps = np.arange(0, 10, 2)
print("\nArange:", steps)

# Linspace (Start, Stop INCLUSIVE, Number of points)
points = np.linspace(0, 100, 5)
print("\nLinspace:", points)

# Random generation (Uniform 0-1)
rands = np.random.rand(3, 3)
print("\nRandoms:\n", rands)
````

<h2 style="color: #000000ff; background-color: #02fff2ff; border-left: 6px solid #0c0c0cff; padding: 12px 15px; margin-top: 40px; border-radius: 4px;">
    <b>3. Vectorized Operations : <span style="color: #070707ff; font-size: 0.85em; font-weight: normal;">No More For-Loops</span></b>
</h2>

<div style="border-left: 4px solid #4DB6AC; padding-left: 15px; margin: 15px 0; color: #CCCCCC;">
    <b>🔍 What is it?</b><br>
    The most important concept in NumPy is <b>Vectorization</b>. You can perform math on entire arrays at once without writing <code>for</code> loops. This is executed in optimized C code, making it blazing fast.
</div>

```python
# ❌ Bad (Python Loop)
for i in range(len(arr)): arr[i] *= 2

# ✅ Good (Vectorized)
arr = arr * 2
```

<h3 style="color: #64B5F6; background-color: #0F1A2A; padding: 8px 12px; border-radius: 4px; margin-top: 25px;"><b>💼 Why Data Analysts Care</b></h3>

* **Financial Math**: Applying a 5% interest rate to a million bank accounts instantly (`balances * 1.05`)
* **Image Processing**: Brightening an image matrix by adding a constant (`pixels + 50`)

<h3 style="color: #81C784; background-color: #102414; padding: 8px 12px; border-radius: 4px; margin-top: 25px;"><b>🧠 Pro Tip</b></h3>

<div style="border-left: 4px solid #81C784; padding-left: 15px; margin: 15px 0; color: #CCCCCC;">
    Arrays operate element-wise. If you add two arrays `A + B`, they must be the exact same shape (or compatible via 'broadcasting'). They add position 0 to position 0, 1 to 1, etc.
</div>

````python
import numpy as np
import time

arr = np.array([1, 2, 3, 4, 5])

# Scalar Math (Applies to all elements)
print(f"Add: {arr + 10}")
print(f"Multiply: {arr * 10}")
print(f"Power: {arr ** 2}")

# Array to Array Math (Element-wise)
arr2 = np.array([10, 20, 30, 40, 50])
print(f"\nArray Addition: {arr + arr2}")

# Speed test!
big_list = list(range(1000000))
big_arr = np.array(big_list)

t1 = time.time()
_ = [x * 2 for x in big_list]  # Python loop
list_time = time.time() - t1

t2 = time.time()
_ = big_arr * 2  # NumPy Vectorization
arr_time = time.time() - t2

print(f"\nNumPy is {list_time / arr_time:.1f}x faster!")
````
