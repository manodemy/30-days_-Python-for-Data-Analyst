<h2 style="color: #000000ff; background-color: #02fff2ff; border-left: 6px solid #0c0c0cff; padding: 12px 15px; margin-top: 40px; border-radius: 4px;">
    <b>1. Dictionary Creation & Basics : <span style="color: #070707ff; font-size: 0.85em; font-weight: normal;">Key-Value Mapping</span></b>
</h2>

<div style="border-left: 4px solid #4DB6AC; padding-left: 15px; margin: 15px 0; color: #CCCCCC;">
    <b>🔍 What is it?</b><br>
    A dictionary is a <b>mutable, unordered collection of key-value pairs</b>. It is Python's most powerful data structure, serving as the foundation for JSON parsing, DataFrames, and fast lookups. Keys must be <b>hashable</b> (strings, numbers, tuples), while values can be anything.
</div>

| Creation Method | Example | Note |
| :--- | :--- | :--- |
| Braces | `{'a': 1, 'b': 2}` | Standard literal |
| `dict()` constructor | `dict(a=1, b=2)` | Clean for string keys |
| From pairs | `dict([('a',1), ('b',2)])`| Converting lists of tuples |
| `fromkeys()` | `dict.fromkeys(['a','b'], 0)`| Initialize with defaults |
| Empty dict | `{}` or `dict()` | Empty dictionary |

<h3 style="color: #64B5F6; background-color: #0F1A2A; padding: 8px 12px; border-radius: 4px; margin-top: 25px;"><b>💼 Why Data Analysts Care</b></h3>

* **JSON Data**: Dictionaries are the Python equivalent of JSON objects
* **Fast Lookups**: Looking up a value by its key is an O(1) operation
* **Data Modeling**: Representing real-world entities (e.g., a user profile)

<h3 style="color: #FFB74D; background-color: #2B1A0D; padding: 8px 12px; border-radius: 4px; margin-top: 25px;"><b>⚠️ Mutable Keys</b></h3>

<div style="border-left: 4px solid #FFB74D; padding-left: 15px; margin: 15px 0; color: #CCCCCC;">
    You cannot use a list or dictionary as a dictionary key because they are unhashable (mutable). You will get a <code>TypeError: unhashable type: 'list'</code>. Use tuples instead.
</div>

<h3 style="color: #81C784; background-color: #102414; padding: 8px 12px; border-radius: 4px; margin-top: 25px;"><b>🧠 Pro Tip</b></h3>

<div style="border-left: 4px solid #81C784; padding-left: 15px; margin: 15px 0; color: #CCCCCC;">
    Since Python 3.7, dictionaries maintain <b>insertion order</b>. But never rely on this for logic; if order matters explicitly, document it or use <code>collections.OrderedDict</code> for backward compatibility.
</div>

````python
# Creation methods
user = {
    'id': 101,
    'name': 'Alice',
    'active': True
}

# dict() constructor (keys become strings automatically)
config = dict(host='localhost', port=8080, debug=False)

# fromkeys for initialization
metrics = dict.fromkeys(['clicks', 'views', 'sales'], 0)

print(f'User: {user}')
print(f'Config: {config}')
print(f'Metrics: {metrics}')

# Fast lookup
print(f'User Name: {user["name"]}')
````

<h2 style="color: #000000ff; background-color: #02fff2ff; border-left: 6px solid #0c0c0cff; padding: 12px 15px; margin-top: 40px; border-radius: 4px;">
    <b>2. Accessing & Modifying Data : <span style="color: #070707ff; font-size: 0.85em; font-weight: normal;">Safe Lookups & Updates</span></b>
</h2>

<div style="border-left: 4px solid #4DB6AC; padding-left: 15px; margin: 15px 0; color: #CCCCCC;">
    <b>🔍 What is it?</b><br>
    You can access values using <code>dict[key]</code>, but if the key doesn't exist, it raises a <code>KeyError</code>. To safely access data, use the <b><code>.get(key, default)</code></b> method. To add or modify, simply assign <code>dict[key] = value</code>.
</div>

| Operation | Syntax | Behavior |
| :--- | :--- | :--- |
| Unsafe Access | `d['key']` | Returns value or raises `KeyError` |
| Safe Access | `d.get('key')` | Returns value or `None` (no error) |
| Safe w/ Default | `d.get('key', 0)` | Returns value or `0` |
| Add / Modify | `d['key'] = val` | Creates new or overwrites existing |
| Delete | `del d['key']` | Removes key-value pair |
| Pop & Return | `d.pop('key')` | Removes and returns the value |

<h3 style="color: #64B5F6; background-color: #0F1A2A; padding: 8px 12px; border-radius: 4px; margin-top: 25px;"><b>💼 Why Data Analysts Care</b></h3>

* **API Parsing**: APIs often have missing fields. Use `.get()` to avoid crashing your pipeline
* **Counter Patterns**: `d[key] = d.get(key, 0) + 1` is the standard way to count frequencies
* **Configuration Defaults**: `port = config.get('port', 80)` applies safe default values

<h3 style="color: #FFB74D; background-color: #2B1A0D; padding: 8px 12px; border-radius: 4px; margin-top: 25px;"><b>⚠️ KeyError Crashes</b></h3>

<div style="border-left: 4px solid #FFB74D; padding-left: 15px; margin: 15px 0; color: #CCCCCC;">
    <code>data['missing_key']</code> is the #1 cause of crashes in data pipelines. Always use <code>.get()</code> when parsing external data.
</div>

````python
record = {'id': 1, 'name': 'DataPipeline'}

# Adding and Modifying
record['status'] = 'running'  # Add new
record['id'] = 2              # Overwrite
print(f'Updated: {record}')

# Safe access
# print(record['owner'])      # CRASH! KeyError
print(record.get('owner'))    # None
print(record.get('owner', 'Admin'))  # 'Admin' (fallback)

# Deleting
del record['status']
removed_id = record.pop('id')
print(f'Popped ID: {removed_id}')
print(f'Final: {record}')
````

<h2 style="color: #000000ff; background-color: #02fff2ff; border-left: 6px solid #0c0c0cff; padding: 12px 15px; margin-top: 40px; border-radius: 4px;">
    <b>3. Dictionary Views : <span style="color: #070707ff; font-size: 0.85em; font-weight: normal;">Keys, Values, and Items</span></b>
</h2>

<div style="border-left: 4px solid #4DB6AC; padding-left: 15px; margin: 15px 0; color: #CCCCCC;">
    <b>🔍 What is it?</b><br>
    Dictionaries provide dynamic <b>view objects</b> through three methods: <code>.keys()</code>, <code>.values()</code>, and <code>.items()</code>. These views update automatically when the dictionary changes and are fast and memory-efficient.
</div>

| Method | Returns View Of | Example Use |
| :--- | :--- | :--- |
| `.keys()` | All keys | Checking membership: `'age' in d.keys()` |
| `.values()` | All values | Statistics: `sum(d.values())` |
| `.items()` | `(key, value)` tuples | Iteration: `for k, v in d.items():` |

<h3 style="color: #64B5F6; background-color: #0F1A2A; padding: 8px 12px; border-radius: 4px; margin-top: 25px;"><b>💼 Why Data Analysts Care</b></h3>

* **Column Checks**: Check if expected columns exist in a data record using `.keys()`
* **Data Extraction**: Extract all metric values for analysis using `.values()`
* **Transformation**: Iterate over `.items()` to apply a function to all values

<h3 style="color: #FFB74D; background-color: #2B1A0D; padding: 8px 12px; border-radius: 4px; margin-top: 25px;"><b>⚠️ Views are not Lists</b></h3>

<div style="border-left: 4px solid #FFB74D; padding-left: 15px; margin: 15px 0; color: #CCCCCC;">
    Views don't support indexing. <code>d.keys()[0]</code> raises a <code>TypeError</code>. If you need a list, explicitly convert it: <code>list(d.keys())</code>.
</div>

````python
scores = {'math': 95, 'physics': 88, 'chem': 92}

# View objects
keys = scores.keys()
vals = scores.values()
items = scores.items()

print(f'Keys: {keys}')
print(f'Values: {vals}')
print(f'Items: {items}')

# Views are dynamic!
scores['biology'] = 90
print(f'\nAfter adding biology, keys view auto-updated: \n{keys}')

# Need a list? Convert it explicitly
key_list = list(keys)
print(f'First key: {key_list[0]}')
````

<h2 style="color: #000000ff; background-color: #02fff2ff; border-left: 6px solid #0c0c0cff; padding: 12px 15px; margin-top: 40px; border-radius: 4px;">
    <b>4. Iteration Patterns : <span style="color: #070707ff; font-size: 0.85em; font-weight: normal;">Looping over Dictionaries</span></b>
</h2>

<div style="border-left: 4px solid #4DB6AC; padding-left: 15px; margin: 15px 0; color: #CCCCCC;">
    <b>🔍 What is it?</b><br>
    Iterating directly over a dictionary loops over its <b>keys</b>. To loop over both keys and values, use <code>.items()</code>. This is the most common and Pythonic way to process dictionary data.
</div>

```python
# ❌ Bad: Looping keys to get values
for k in d:
    v = d[k]  # Inefficient lookup

# ✅ Good: Unpacking items directly
for k, v in d.items():
    print(k, v)
```

<h3 style="color: #64B5F6; background-color: #0F1A2A; padding: 8px 12px; border-radius: 4px; margin-top: 25px;"><b>💼 Why Data Analysts Care</b></h3>

* **Formatting Output**: Printing structured reports of metrics or configuration
* **Data Cleaning**: Iterating over items to clean strings or convert types
* **Filtering**: Looping to build a new dictionary with specific key-value pairs

<h3 style="color: #81C784; background-color: #102414; padding: 8px 12px; border-radius: 4px; margin-top: 25px;"><b>🧠 Pro Tip</b></h3>

<div style="border-left: 4px solid #81C784; padding-left: 15px; margin: 15px 0; color: #CCCCCC;">
    You cannot add or remove keys from a dictionary while iterating over it. It will raise a <code>RuntimeError: dictionary changed size during iteration</code>. If you must modify keys, iterate over a copy or build a new dictionary.
</div>

````python
config = {'host': 'localhost', 'port': 5432, 'user': 'admin'}

# Default iteration is keys
print("Keys only:")
for k in config:
    print(f"- {k}")

# Iterating values
print("\nValues only:")
for v in config.values():
    print(f"- {v}")

# The Pythonic Way: items() unpacking
print("\nItems (Keys & Values):")
for key, value in config.items():
    print(f"{key.upper()}: {value}")
````

<h2 style="color: #000000ff; background-color: #02fff2ff; border-left: 6px solid #0c0c0cff; padding: 12px 15px; margin-top: 40px; border-radius: 4px;">
    <b>5. Dictionary Comprehensions : <span style="color: #070707ff; font-size: 0.85em; font-weight: normal;">Functional Dictionary Creation</span></b>
</h2>

<div style="border-left: 4px solid #4DB6AC; padding-left: 15px; margin: 15px 0; color: #CCCCCC;">
    <b>🔍 What is it?</b><br>
    Just like list comprehensions, <b>dictionary comprehensions</b> provide a concise way to create dictionaries from iterables. The syntax is <code>{key: value for item in iterable if condition}</code>.
</div>

```python
# Square numbers mapping
squares = {x: x**2 for x in range(5)}
# {0: 0, 1: 1, 2: 4, 3: 9, 4: 16}
```

<h3 style="color: #64B5F6; background-color: #0F1A2A; padding: 8px 12px; border-radius: 4px; margin-top: 25px;"><b>💼 Why Data Analysts Care</b></h3>

* **Data Transformation**: Convert a list of records into a dictionary keyed by ID for fast lookups
* **Filtering Dictionaries**: `{k: v for k, v in data.items() if v > 0}` — clean negative values
* **Key Mapping**: Mapping original column names to new clean column names

<h3 style="color: #81C784; background-color: #102414; padding: 8px 12px; border-radius: 4px; margin-top: 25px;"><b>🧠 Pro Tip</b></h3>

<div style="border-left: 4px solid #81C784; padding-left: 15px; margin: 15px 0; color: #CCCCCC;">
    A very common pattern for fast lookups: <code>lookup = {row['id']: row for row in dataset}</code>. This converts an O(n) search into an O(1) dictionary lookup.
</div>

````python
names = ['Alice', 'Bob', 'Charlie']

# Create dict mapping name to its length
name_lengths = {name: len(name) for name in names}
print(f'Lengths: {name_lengths}')

# Filtering an existing dictionary
prices = {'apple': 1.2, 'banana': 0.5, 'cherry': 2.5}
expensive = {k: v for k, v in prices.items() if v > 1.0}
print(f'Expensive items: {expensive}')

# Swapping keys and values
original = {'a': 1, 'b': 2, 'c': 3}
swapped = {v: k for k, v in original.items()}
print(f'Swapped: {swapped}')
````

<h2 style="color: #000000ff; background-color: #02fff2ff; border-left: 6px solid #0c0c0cff; padding: 12px 15px; margin-top: 40px; border-radius: 4px;">
    <b>6. Merging & Updating : <span style="color: #070707ff; font-size: 0.85em; font-weight: normal;">Combining Dictionaries</span></b>
</h2>

<div style="border-left: 4px solid #4DB6AC; padding-left: 15px; margin: 15px 0; color: #CCCCCC;">
    <b>🔍 What is it?</b><br>
    Python provides multiple ways to combine dictionaries. <code>dict.update()</code> modifies in-place. Python 3.9+ introduced the <b>merge operator <code>|</code></b> which creates a new dictionary.
</div>

| Operation | Syntax | Effect |
| :--- | :--- | :--- |
| Update in-place | `d1.update(d2)` | `d1` is modified. `d2` keys overwrite `d1`. Returns `None`. |
| Merge operator (3.9+) | `d3 = d1 \| d2` | Creates new dict `d3`. `d2` overwrites `d1`. |
| Unpacking | `d3 = {**d1, **d2}` | Creates new dict. Pre-3.9 equivalent of `\|`. |

<h3 style="color: #64B5F6; background-color: #0F1A2A; padding: 8px 12px; border-radius: 4px; margin-top: 25px;"><b>💼 Why Data Analysts Care</b></h3>

* **Configuration Management**: Merge `default_config` with `user_config` (user overrides default)
* **Aggregating Data**: Combining parsed JSON objects from multiple API endpoints

<h3 style="color: #FFB74D; background-color: #2B1A0D; padding: 8px 12px; border-radius: 4px; margin-top: 25px;"><b>⚠️ Right Wins</b></h3>

<div style="border-left: 4px solid #FFB74D; padding-left: 15px; margin: 15px 0; color: #CCCCCC;">
    When merging, if the same key exists in both dictionaries, the value from the dictionary on the <b>right side</b> (or passed to `.update()`) wins.
</div>

````python
defaults = {'theme': 'light', 'port': 8080, 'debug': False}
user_prefs = {'theme': 'dark', 'verbose': True}

# 1. Update in-place
active_config = defaults.copy()
active_config.update(user_prefs)
print(f'Update: {active_config}')

# 2. Merge operator (Python 3.9+)
merged_config = defaults | user_prefs
print(f'Merge |: {merged_config}')

# 3. Kwargs unpacking (older Python)
unpacked = {**defaults, **user_prefs}
print(f'Unpacked: {unpacked}')
````

    <b>7. Nested Dictionaries : <span style="color: #070707ff; font-size: 0.85em; font-weight: normal;">JSON & Deep Data</span></b>
</h2>

<div style="border-left: 4px solid #4DB6AC; padding-left: 15px; margin: 15px 0; color: #CCCCCC;">
    <b>🔍 What is it?</b><br>
    A dictionary can contain other dictionaries (or lists, sets, etc.) as values. This is essential for working with hierarchical data like <b>JSON</b> or <b>NoSQL document databases</b>.
</div>

```python
users = {
    'u1': {'name': 'Alice', 'role': 'Admin'},
    'u2': {'name': 'Bob', 'role': 'User'}
}
# Access nested data
alice_role = users['u1']['role']
```

<h3 style="color: #64B5F6; background-color: #0F1A2A; padding: 8px 12px; border-radius: 4px; margin-top: 25px;"><b>💼 Why Data Analysts Care</b></h3>

* **API Parsing**: Traversing complex JSON responses from web services
* **Document Storage**: Representing MongoDB documents in memory
* **Configuration Trees**: Hierarchical settings (e.g., `config['database']['host']`)

<h3 style="color: #FFB74D; background-color: #2B1A0D; padding: 8px 12px; border-radius: 4px; margin-top: 25px;"><b>⚠️ Deep Missing Keys</b></h3>

<div style="border-left: 4px solid #FFB74D; padding-left: 15px; margin: 15px 0; color: #CCCCCC;">
    Accessing <code>data['user']['address']['zip']</code> will crash if <i>any</i> key in the chain is missing. Use chained `.get()` or handle exceptions.
</div>

````python
import json

# JSON-like nested structure
api_response = {
    'status': 200,
    'data': {
        'user': {'id': 101, 'name': 'Alice'},
        'permissions': ['read', 'write']
    }
}

# Accessing deep data
user_name = api_response['data']['user']['name']
perms = api_response['data']['permissions']
print(f'User: {user_name}, Perms: {perms}')

# Safe deep access (avoid KeyError)
data_obj = api_response.get('data', {})
address = data_obj.get('user', {}).get('address', 'No Address')
print(f'Address: {address}')

# Pretty printing nested dicts
print("\nPretty printed:")
print(json.dumps(api_response, indent=2))
````
