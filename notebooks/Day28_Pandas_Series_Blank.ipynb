<h2 style="color: #000000ff; background-color: #02fff2ff; border-left: 6px solid #0c0c0cff; padding: 12px 15px; margin-top: 40px; border-radius: 4px;">
    <b>1. String and Datetime Accessors : <span style="color: #070707ff; font-size: 0.85em; font-weight: normal;">.str and .dt</span></b>
</h2>

<div style="border-left: 4px solid #4DB6AC; padding-left: 15px; margin: 15px 0; color: #CCCCCC;">
    <b>🔍 What is it?</b><br>
    Pandas provides specialized methods for specific data types using <b>Accessors</b>. If a column contains strings, use the <code>.str</code> accessor to apply string methods (like `.upper()`). If it contains dates, use <code>.dt</code> to extract the year or month.
</div>

```python
# String Accessor
df['Name'] = df['Name'].str.upper().str.strip()

# Datetime Accessor
df['Year'] = df['Date'].dt.year
```

<h3 style="color: #64B5F6; background-color: #0F1A2A; padding: 8px 12px; border-radius: 4px; margin-top: 25px;"><b>💼 Why Data Analysts Care</b></h3>

* **Text Cleaning**: Removing leading/trailing spaces from a column of messy user inputs
* **Seasonality**: Extracting the `DayOfWeek` from a Timestamp to see if weekend sales are higher

<h3 style="color: #FFB74D; background-color: #2B1A0D; padding: 8px 12px; border-radius: 4px; margin-top: 25px;"><b>⚠️ Type Errors</b></h3>

<div style="border-left: 4px solid #FFB74D; padding-left: 15px; margin: 15px 0; color: #CCCCCC;">
    You cannot use <code>.dt</code> on a string column! You must first convert it: <code>df['Date'] = pd.to_datetime(df['DateString'])</code> before using the accessor.
</div>

````python
import pandas as pd

df = pd.DataFrame({
    'Product': [' Apple ', 'BANANA', 'cherry_pie'],
    'Bought': ['2023-12-25', '2024-01-01', '2024-02-14']
})

# String Accessors
df['Clean_Product'] = df['Product'].str.strip().str.title()
print("Clean Strings:\n", df[['Product', 'Clean_Product']])

# Convert to Datetime FIRST
df['Date'] = pd.to_datetime(df['Bought'])

# Datetime Accessors
df['Year'] = df['Date'].dt.year
df['Month'] = df['Date'].dt.month_name()
print("\nDates Extracted:\n", df[['Bought', 'Year', 'Month']])
````

<h2 style="color: #000000ff; background-color: #02fff2ff; border-left: 6px solid #0c0c0cff; padding: 12px 15px; margin-top: 40px; border-radius: 4px;">
    <b>2. Time Series Resampling : <span style="color: #070707ff; font-size: 0.85em; font-weight: normal;">Rolling up Dates</span></b>
</h2>

<div style="border-left: 4px solid #4DB6AC; padding-left: 15px; margin: 15px 0; color: #CCCCCC;">
    <b>🔍 What is it?</b><br>
    When working with Time Series data (where the <b>Index</b> is a Datetime), you can use <b><code>.resample()</code></b>. It is exactly like <code>.groupby()</code>, but specifically for time frequencies (e.g., grouping daily data into monthly averages).
</div>

| Frequency | String | Example |
| :--- | :--- | :--- |
| Daily | `'D'` | `df.resample('D').sum()` |
| Weekly | `'W'` | `df.resample('W').mean()` |
| Monthly | `'M'` | `df.resample('M').max()` |
| Quarterly| `'Q'` | `df.resample('Q').count()` |

<h3 style="color: #64B5F6; background-color: #0F1A2A; padding: 8px 12px; border-radius: 4px; margin-top: 25px;"><b>💼 Why Data Analysts Care</b></h3>

* **Financial Aggregations**: Converting minute-by-tick stock data into Daily OHLC (Open, High, Low, Close) bars
* **Smoothing**: Aggregating noisy daily web traffic into smooth Weekly averages

<h3 style="color: #81C784; background-color: #102414; padding: 8px 12px; border-radius: 4px; margin-top: 25px;"><b>🧠 Pro Tip</b></h3>

<div style="border-left: 4px solid #81C784; padding-left: 15px; margin: 15px 0; color: #CCCCCC;">
    <code>.resample()</code> ONLY works if your DataFrame's index is a <code>DatetimeIndex</code>! Use <code>df.set_index('DateColumn', inplace=True)</code> before attempting to resample.
</div>

````python
import pandas as pd
import numpy as np

# Create 10 days of data
dates = pd.date_range(start='2024-01-01', periods=10, freq='D')
df = pd.DataFrame({
    'Sales': np.random.randint(100, 500, 10)
}, index=dates)

print("Daily Data (First 5):\n", df.head())

# Resample to Weekly ('W') and Sum
weekly = df.resample('W').sum()
print("\nWeekly Total Sales:\n", weekly)

# Resample to 3-Day periods and Mean
threeday = df.resample('3D').mean()
print("\n3-Day Average Sales:\n", threeday)
````

<h2 style="color: #000000ff; background-color: #02fff2ff; border-left: 6px solid #0c0c0cff; padding: 12px 15px; margin-top: 40px; border-radius: 4px;">
    <b>3. Rolling Windows : <span style="color: #070707ff; font-size: 0.85em; font-weight: normal;">Moving Averages</span></b>
</h2>

<div style="border-left: 4px solid #4DB6AC; padding-left: 15px; margin: 15px 0; color: #CCCCCC;">
    <b>🔍 What is it?</b><br>
    A <b>Rolling Window</b> calculates statistics over a sliding window of time. The most common use case is the <b>Moving Average</b>, which smooths out short-term fluctuations in data to highlight longer-term trends.
</div>

```python
# Calculate a 7-day moving average
df['7D_MA'] = df['Sales'].rolling(window=7).mean()
```

<h3 style="color: #64B5F6; background-color: #0F1A2A; padding: 8px 12px; border-radius: 4px; margin-top: 25px;"><b>💼 Why Data Analysts Care</b></h3>

* **Stock Analysis**: Calculating the 50-day and 200-day moving averages to find 'Golden Cross' buy signals
* **Trend Detection**: Smoothing out weekday vs weekend sales spikes to see the true month-over-month growth

<h3 style="color: #81C784; background-color: #102414; padding: 8px 12px; border-radius: 4px; margin-top: 25px;"><b>🧠 Pro Tip</b></h3>

<div style="border-left: 4px solid #81C784; padding-left: 15px; margin: 15px 0; color: #CCCCCC;">
    The first <code>N-1</code> rows of a rolling window of size <code>N</code> will result in <code>NaN</code>, because there isn't enough historical data to calculate the full window yet. You can use <code>min_periods=1</code> to calculate partial windows.
</div>

````python
import pandas as pd

# 10 days of volatile sales data
df = pd.DataFrame({
    'Sales': [10, 100, 20, 90, 15, 110, 25, 95, 30, 105]
})

# Calculate a 3-day Rolling Mean
df['3D_Avg'] = df['Sales'].rolling(window=3).mean()

# Calculate a 3-day Rolling Sum
df['3D_Sum'] = df['Sales'].rolling(window=3).sum()

print("Notice the NaNs in the first 2 rows!")
print(df)
````
