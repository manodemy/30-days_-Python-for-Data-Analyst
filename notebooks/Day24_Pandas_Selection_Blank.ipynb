<h2 style="color: #000000ff; background-color: #02fff2ff; border-left: 6px solid #0c0c0cff; padding: 12px 15px; margin-top: 40px; border-radius: 4px;">
    <b>1. Loc and Iloc : <span style="color: #070707ff; font-size: 0.85em; font-weight: normal;">Row and Column Selection</span></b>
</h2>

<div style="border-left: 4px solid #4DB6AC; padding-left: 15px; margin: 15px 0; color: #CCCCCC;">
    <b>🔍 What is it?</b><br>
    Selecting data in Pandas requires precision. <b><code>.iloc[]</code></b> selects by integer position (like standard Python lists). <b><code>.loc[]</code></b> selects by index label and column name. Both follow the format <code>[rows, columns]</code>.
</div>

| Accessor | Paradigm | Example | Meaning |
| :--- | :--- | :--- | :--- |
| `.iloc` | Integer Position | `df.iloc[0:5, 0:2]` | First 5 rows, first 2 columns |
| `.loc` | Label / Name | `df.loc[:, 'Age':'City']` | All rows, columns from Age to City |

<h3 style="color: #64B5F6; background-color: #0F1A2A; padding: 8px 12px; border-radius: 4px; margin-top: 25px;"><b>💼 Why Data Analysts Care</b></h3>

* **Feature Selection**: Extracting the target variable `Y = df['Price']` and the features `X = df.iloc[:, :-1]` for Machine Learning

<h3 style="color: #FFB74D; background-color: #2B1A0D; padding: 8px 12px; border-radius: 4px; margin-top: 25px;"><b>⚠️ Inclusive Loc</b></h3>

<div style="border-left: 4px solid #FFB74D; padding-left: 15px; margin: 15px 0; color: #CCCCCC;">
    A major 'gotcha': Slicing with <code>.iloc[0:5]</code> is EXCLUSIVE of 5 (returns 0,1,2,3,4). But slicing with <code>.loc['A':'C']</code> is INCLUSIVE of 'C' (returns A,B,C).
</div>

````python
import pandas as pd

df = pd.DataFrame({
    'Name': ['Alice', 'Bob', 'Charlie', 'David'],
    'Age': [25, 30, 35, 40],
    'Salary': [50000, 60000, 70000, 80000]
}, index=['ID1', 'ID2', 'ID3', 'ID4'])

# iloc: Integer position (Row 1, all columns)
print("iloc (Position):\n", df.iloc[1, :])

# loc: Label based (Row 'ID2', specific columns)
print("\nloc (Label):\n", df.loc['ID2', ['Name', 'Salary']])

# Slicing with loc (Inclusive!)
print("\nloc Slicing:\n", df.loc['ID1':'ID3', 'Age':])
````

<h2 style="color: #000000ff; background-color: #02fff2ff; border-left: 6px solid #0c0c0cff; padding: 12px 15px; margin-top: 40px; border-radius: 4px;">
    <b>2. Boolean Filtering : <span style="color: #070707ff; font-size: 0.85em; font-weight: normal;">Querying Data</span></b>
</h2>

<div style="border-left: 4px solid #4DB6AC; padding-left: 15px; margin: 15px 0; color: #CCCCCC;">
    <b>🔍 What is it?</b><br>
    You can filter DataFrames exactly like NumPy arrays using Boolean Masks. Create a condition, and pass it into the DataFrame brackets: <code>df[condition]</code>. Pandas also provides the highly readable <code>.query()</code> method for SQL-like string queries.
</div>

```python
# Standard Masking
rich_users = df[df['Salary'] > 75000]

# Multiple conditions (Requires parenthesis and bitwise &)
targets = df[(df['Age'] > 30) & (df['City'] == 'NY')]

# The elegant .query() alternative
targets = df.query("Age > 30 and City == 'NY'")
```

<h3 style="color: #64B5F6; background-color: #0F1A2A; padding: 8px 12px; border-radius: 4px; margin-top: 25px;"><b>💼 Why Data Analysts Care</b></h3>

* **Cohort Extraction**: Filtering a massive dataset down to only active users from a specific country
* **Date Filtering**: `df[df['Date'] >= '2024-01-01']` to extract year-to-date performance

<h3 style="color: #81C784; background-color: #102414; padding: 8px 12px; border-radius: 4px; margin-top: 25px;"><b>🧠 Pro Tip</b></h3>

<div style="border-left: 4px solid #81C784; padding-left: 15px; margin: 15px 0; color: #CCCCCC;">
    When using <code>.query()</code>, you can refer to external Python variables by prefixing them with an <code>@</code> symbol: <code>df.query('Age > @min_age')</code>.
</div>

````python
import pandas as pd

df = pd.DataFrame({
    'Name': ['Alice', 'Bob', 'Charlie', 'David'],
    'Age': [25, 30, 35, 40],
    'Dept': ['HR', 'IT', 'IT', 'Sales']
})

# Boolean Mask Filtering
it_staff = df[df['Dept'] == 'IT']
print("Mask Filtering:\n", it_staff)

# .isin() for multiple values
hr_or_sales = df[df['Dept'].isin(['HR', 'Sales'])]
print("\n.isin() Filtering:\n", hr_or_sales)

# Query syntax (Very readable)
senior_it = df.query("Dept == 'IT' and Age >= 35")
print("\n.query() Filtering:\n", senior_it)
````

<h2 style="color: #000000ff; background-color: #02fff2ff; border-left: 6px solid #0c0c0cff; padding: 12px 15px; margin-top: 40px; border-radius: 4px;">
    <b>3. Sorting and Ranking : <span style="color: #070707ff; font-size: 0.85em; font-weight: normal;">Ordering Data</span></b>
</h2>

<div style="border-left: 4px solid #4DB6AC; padding-left: 15px; margin: 15px 0; color: #CCCCCC;">
    <b>🔍 What is it?</b><br>
    Once data is filtered, we often need to sort it to find the top/bottom performers using <code>.sort_values()</code>. You can sort by multiple columns and specify ascending/descending order.
</div>

| Method | Purpose | Example |
| :--- | :--- | :--- |
| `sort_values(by=)` | Sort rows | `df.sort_values(by='Age', ascending=False)` |
| `nlargest(n, col)` | Top N rows | `df.nlargest(5, 'Salary')` (Faster than sorting!) |
| `rank()` | Assign ranks | `df['Salary'].rank(ascending=False)` |

<h3 style="color: #64B5F6; background-color: #0F1A2A; padding: 8px 12px; border-radius: 4px; margin-top: 25px;"><b>💼 Why Data Analysts Care</b></h3>

* **Leaderboards**: Finding the top 10 highest revenue generating products
* **Time Series**: Ensuring financial data is strictly sorted chronologically before calculating moving averages

<h3 style="color: #81C784; background-color: #102414; padding: 8px 12px; border-radius: 4px; margin-top: 25px;"><b>🧠 Pro Tip</b></h3>

<div style="border-left: 4px solid #81C784; padding-left: 15px; margin: 15px 0; color: #CCCCCC;">
    If you just need the top 5 values, <code>df.nlargest(5, 'Col')</code> is computationally faster than sorting the entire million-row DataFrame and calling <code>.head(5)</code>.
</div>

````python
import pandas as pd

df = pd.DataFrame({
    'Name': ['Alice', 'Bob', 'Charlie', 'David'],
    'Score': [85, 92, 85, 95],
    'Time': [120, 110, 105, 130]
})

# Sort by single column (Descending)
print("Highest Scores:\n", df.sort_values(by='Score', ascending=False))

# Sort by multiple columns (Score desc, then Time asc to break ties)
tie_breaker = df.sort_values(by=['Score', 'Time'], ascending=[False, True])
print("\nMulti-sort (Tiebreaker):\n", tie_breaker)

# Find top 2 using nlargest
print("\nTop 2 Scores:\n", df.nlargest(2, 'Score'))
````
