    A subquery is a nested query. Scalar subqueries return a single value. Multi-row subqueries return a list of values (used with IN, NOT IN, EXISTS). Correlated subqueries reference columns in the outer query and execute once per candidate row. SQLite supports subqueries across all clauses.
