    Views are saved query blocks served as virtual tables. Stored procedures accept inputs and wrap transaction logic (SQLite does not support stored procedures; standard databases do). Triggers execute automatic logic in response to INSERT, UPDATE, or DELETE statements.

```sql
CREATE VIEW IF NOT EXISTS v_dept_summary AS SELECT d.name, COUNT(e.id) as staff_count FROM departments d LEFT JOIN employees e ON d.id = e.dept_id GROUP BY d.id;
```
