<div style="border-left: 4px solid #4DB6AC; padding-left: 15px; margin: 15px 0; color: #CCCCCC;">
    <b>🔍 Concept</b><br>
    Window functions perform calculations across a set of rows related to the current row, without collapsing them. The OVER clause configures this partition window. ROW_NUMBER assigns sequential numbers. RANK skips numbers on duplicate ties, whereas DENSE_RANK assigns consecutive numbers.
</div>

```sql
SELECT name, category_id, cost_price, DENSE_RANK() OVER(PARTITION BY category_id ORDER BY cost_price DESC) as price_rank FROM products;
```
