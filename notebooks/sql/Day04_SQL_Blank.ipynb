<div style="border-left: 4px solid #4DB6AC; padding-left: 15px; margin: 15px 0; color: #CCCCCC;">
    <b>🔍 Concept</b><br>
    Aggregate functions (COUNT, SUM, AVG, MIN, MAX) perform calculations on columns, ignoring NULL values (except COUNT(*)). GROUP BY partitions rows into summary buckets. The HAVING clause filters these grouped summaries based on aggregate conditions. WHERE filters rows before aggregation occurs.
</div>

```sql
SELECT category, COUNT(*) as orders, AVG(amount) as avg_sale FROM sales GROUP BY category HAVING avg_sale > 3000;
```
